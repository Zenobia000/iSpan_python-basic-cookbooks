````python
# ç·´ç¿’ 1 è§£ç­”: æ•¸å­—ç¸½å’Œ
def sum_to_n(n):
    """è¨ˆç®—å¾ 1 åˆ° n çš„ç¸½å’Œ"""
    # åŸºæœ¬æƒ…æ³ï¼šn = 1 æ™‚å›å‚³ 1
    if n == 1:
        return 1
    
    # éè¿´æƒ…æ³ï¼šn + sum_to_n(n-1)
    return n + sum_to_n(n - 1)

# æ¸¬è©¦
print("ç·´ç¿’ 1 æ¸¬è©¦çµæœï¼š")
test_cases = [(1, 1), (3, 6), (5, 15), (10, 55)]
for n, expected in test_cases:
    result = sum_to_n(n)
    print(f"sum_to_n({n}) = {result} {'âœ…' if result == expected else 'âŒ'}")

# è§£é¡Œæ€è·¯ï¼š
# 1. åŸºæœ¬æƒ…æ³ï¼šn = 1 æ™‚ï¼Œç¸½å’Œç‚º 1
# 2. éè¿´æƒ…æ³ï¼šç•¶å‰æ•¸å­— + å‰©é¤˜æ•¸å­—çš„ç¸½å’Œ
# 3. æ™‚é–“è¤‡é›œåº¦ï¼šO(n)ï¼Œç©ºé–“è¤‡é›œåº¦ï¼šO(n)ï¼ˆéè¿´æ·±åº¦ï¼‰
````

````python
# ç·´ç¿’ 2 è§£ç­”: æ¬¡æ–¹è¨ˆç®—
def power(base, exp):
    """è¨ˆç®— base çš„ exp æ¬¡æ–¹"""
    # åŸºæœ¬æƒ…æ³ï¼šä»»ä½•æ•¸çš„ 0 æ¬¡æ–¹éƒ½æ˜¯ 1
    if exp == 0:
        return 1
    
    # éè¿´æƒ…æ³ï¼šbase * power(base, exp-1)
    return base * power(base, exp - 1)

# æ¸¬è©¦
print("ç·´ç¿’ 2 æ¸¬è©¦çµæœï¼š")
test_cases = [(2, 0, 1), (2, 3, 8), (5, 2, 25)]
for base, exp, expected in test_cases:
    result = power(base, exp)
    print(f"power({base}, {exp}) = {result} {'âœ…' if result == expected else 'âŒ'}")
````

````python
# ç·´ç¿’ 4 è§£ç­”: æ•¸å­—åè½‰
def reverse_number(n):
    """éè¿´åè½‰æ•´æ•¸çš„æ•¸å­—é †åº"""
    def get_digit_count(num):
        """è¨ˆç®—æ•¸å­—ä½æ•¸"""
        if num < 10:
            return 1
        return 1 + get_digit_count(num // 10)
    
    def reverse_helper(num, power):
        """éè¿´åè½‰è¼”åŠ©å‡½å¼"""
        # åŸºæœ¬æƒ…æ³ï¼šåªæœ‰ä¸€ä½æ•¸
        if num < 10:
            return num
        
        # éè¿´æƒ…æ³ï¼šæœ€å¾Œä¸€ä½ * 10^(ä½æ•¸-1) + åè½‰å‰©é¤˜éƒ¨åˆ†
        last_digit = num % 10
        remaining = num // 10
        return last_digit * (10 ** (power - 1)) + reverse_helper(remaining, power - 1)
    
    if n < 10:
        return n
    
    digit_count = get_digit_count(n)
    return reverse_helper(n, digit_count)

# æ¸¬è©¦
print("ç·´ç¿’ 4 æ¸¬è©¦çµæœï¼š")
test_cases = [(7, 7), (45, 54), (123, 321)]
for n, expected in test_cases:
    result = reverse_number(n)
    print(f"reverse_number({n}) = {result} {'âœ…' if result == expected else 'âŒ'}")
````

````python
# ç¿’é¡Œ 2 è§£ç­”: éè¿´æª¢æŸ¥å›æ–‡æ•¸å­—
def is_palindrome_number(n: int) -> bool:
    """éè¿´æª¢æŸ¥æ­£æ•´æ•¸æ˜¯å¦ç‚ºå›æ–‡æ•¸å­—"""
    def reverse_number_helper(num, reversed_num=0):
        """è¼”åŠ©å‡½å¼ï¼šåè½‰æ•¸å­—"""
        if num == 0:
            return reversed_num
        return reverse_number_helper(num // 10, reversed_num * 10 + num % 10)
    
    # å°‡æ•¸å­—åè½‰å¾Œèˆ‡åŸæ•¸å­—æ¯”è¼ƒ
    return n == reverse_number_helper(n)

# æ¸¬è©¦
print("\nç¿’é¡Œ 2 æ¸¬è©¦ï¼š")
print(f"is_palindrome_number(121) = {is_palindrome_number(121)} (æœŸæœ›: True)")
print(f"is_palindrome_number(123) = {is_palindrome_number(123)} (æœŸæœ›: False)")
````

````python
# ç¿’é¡Œ 3 è§£ç­”: éè¿´è¨ˆç®—é™£åˆ—æœ€å¤§å€¼
def find_max_recursive(arr: list) -> int:
    """éè¿´æ‰¾å‡ºé™£åˆ—ä¸­çš„æœ€å¤§å€¼"""
    # åŸºæœ¬æƒ…æ³ï¼šåªæœ‰ä¸€å€‹å…ƒç´ 
    if len(arr) == 1:
        return arr[0]
    
    # éè¿´æƒ…æ³ï¼šæ¯”è¼ƒç¬¬ä¸€å€‹å…ƒç´ èˆ‡å‰©é¤˜éƒ¨åˆ†çš„æœ€å¤§å€¼
    max_of_rest = find_max_recursive(arr[1:])
    return arr[0] if arr[0] > max_of_rest else max_of_rest

# ç¿’é¡Œ 4 è§£ç­”: éè¿´è¨ˆç®— GCD
def gcd_recursive(a: int, b: int) -> int:
    """ä½¿ç”¨æ­å¹¾é‡Œå¾—æ¼”ç®—æ³•éè¿´è¨ˆç®—æœ€å¤§å…¬å› æ•¸"""
    # åŸºæœ¬æƒ…æ³ï¼šb = 0
    if b == 0:
        return a
    
    # éè¿´æƒ…æ³ï¼šgcd(b, a % b)
    return gcd_recursive(b, a % b)

# ç¿’é¡Œ 5 è§£ç­”: éè¿´è¨ˆç®—é™£åˆ—ä¹˜ç©
def array_product(arr: list) -> int:
    """éè¿´è¨ˆç®—é™£åˆ—ä¸­æ‰€æœ‰å…ƒç´ çš„ä¹˜ç©"""
    # åŸºæœ¬æƒ…æ³ï¼šç©ºé™£åˆ—
    if not arr:
        return 1
    
    # éè¿´æƒ…æ³ï¼šç¬¬ä¸€å€‹å…ƒç´  * å‰©é¤˜å…ƒç´ çš„ä¹˜ç©
    return arr[0] * array_product(arr[1:])

# ç¿’é¡Œ 6 è§£ç­”: éè¿´ç”¢ç”Ÿæ•¸å­—åºåˆ—
def generate_sequence(n: int) -> list:
    """éè¿´ç”¢ç”Ÿå¾ 1 åˆ° n çš„æ•¸å­—åºåˆ—"""
    # åŸºæœ¬æƒ…æ³ï¼šn = 0
    if n == 0:
        return []
    
    # éè¿´æƒ…æ³ï¼šå‰é¢çš„åºåˆ— + ç•¶å‰æ•¸å­—
    return generate_sequence(n - 1) + [n]

print("\nåŸºç¤é¡Œ 3-6 è§£ç­”å·²å®Œæˆ")
````

````python
# ç¿’é¡Œ 7 è§£ç­”: éè¿´å¯¦ä½œäºŒåˆ†æœå°‹
def binary_search_recursive(arr: list, target: int) -> int:
    """éè¿´å¯¦ä½œäºŒåˆ†æœå°‹æ¼”ç®—æ³•"""
    def search_helper(left: int, right: int) -> int:
        # åŸºæœ¬æƒ…æ³ï¼šæœå°‹ç¯„åœç„¡æ•ˆ
        if left > right:
            return -1
        
        mid = (left + right) // 2
        
        # æ‰¾åˆ°ç›®æ¨™
        if arr[mid] == target:
            return mid
        # åœ¨å·¦åŠéƒ¨æœå°‹
        elif arr[mid] > target:
            return search_helper(left, mid - 1)
        # åœ¨å³åŠéƒ¨æœå°‹
        else:
            return search_helper(mid + 1, right)
    
    return search_helper(0, len(arr) - 1)

# æ¸¬è©¦
print("ç¿’é¡Œ 7 æ¸¬è©¦ï¼š")
arr = [1, 3, 5, 7, 9]
print(f"binary_search_recursive({arr}, 5) = {binary_search_recursive(arr, 5)} (æœŸæœ›: 2)")
````

````python
# ç¿’é¡Œ 8 è§£ç­”: éè¿´æª¢æŸ¥é™£åˆ—æ’åº
def is_sorted_recursive(arr: list) -> bool:
    """éè¿´æª¢æŸ¥é™£åˆ—æ˜¯å¦æŒ‰éå¢é †åºæ’åˆ—"""
    # åŸºæœ¬æƒ…æ³ï¼šç©ºé™£åˆ—æˆ–å–®å…ƒç´ é™£åˆ—
    if len(arr) <= 1:
        return True
    
    # æª¢æŸ¥å‰å…©å€‹å…ƒç´ çš„é †åºï¼Œä¸¦éè¿´æª¢æŸ¥å‰©é¤˜éƒ¨åˆ†
    if arr[0] > arr[1]:
        return False
    
    return is_sorted_recursive(arr[1:])

# ç¿’é¡Œ 9 è§£ç­”: éè¿´è¨ˆç®—æ¨¹ç‹€çµæ§‹æ·±åº¦
def max_depth(nested_list) -> int:
    """éè¿´è¨ˆç®—å·¢ç‹€åˆ—è¡¨çš„æœ€å¤§æ·±åº¦"""
    # åŸºæœ¬æƒ…æ³ï¼šç©ºåˆ—è¡¨
    if not nested_list:
        return 0
    
    max_sub_depth = 0
    for item in nested_list:
        if isinstance(item, list):
            sub_depth = max_depth(item)
            max_sub_depth = max(max_sub_depth, sub_depth)
    
    return 1 + max_sub_depth

print("\nä¸­ç­‰é¡Œ 8-9 è§£ç­”å·²å®Œæˆ")
````

````python
# ç¿’é¡Œ 10 è§£ç­”: éè¿´ç”¢ç”Ÿ Pascal ä¸‰è§’å½¢
def pascal_triangle_row(n: int) -> list:
    """éè¿´ç”¢ç”Ÿ Pascal ä¸‰è§’å½¢çš„ç¬¬ n è¡Œ"""
    # åŸºæœ¬æƒ…æ³ï¼šç¬¬ 0 è¡Œ
    if n == 0:
        return [1]
    
    # å–å¾—ä¸Šä¸€è¡Œ
    prev_row = pascal_triangle_row(n - 1)
    
    # å»ºæ§‹ç•¶å‰è¡Œ
    current_row = [1]  # æ¯è¡Œéƒ½ä»¥ 1 é–‹å§‹
    
    # ä¸­é–“å…ƒç´  = ä¸Šä¸€è¡Œç›¸é„°å…©å…ƒç´ ä¹‹å’Œ
    for i in range(1, len(prev_row)):
        current_row.append(prev_row[i-1] + prev_row[i])
    
    current_row.append(1)  # æ¯è¡Œéƒ½ä»¥ 1 çµæŸ
    
    return current_row

# ç¿’é¡Œ 11 è§£ç­”: éè¿´è¨ˆç®—çµ„åˆæ•¸
def combination_recursive(n: int, r: int) -> int:
    """éè¿´è¨ˆç®—çµ„åˆæ•¸ C(n, r)"""
    # åŸºæœ¬æƒ…æ³
    if r == 0 or r == n:
        return 1
    
    # éè¿´æƒ…æ³ï¼šC(n, r) = C(n-1, r-1) + C(n-1, r)
    return combination_recursive(n - 1, r - 1) + combination_recursive(n - 1, r)

# ç¿’é¡Œ 12 è§£ç­”: éè¿´å­—ä¸²æ›¿æ›
def replace_char_recursive(s: str, old_char: str, new_char: str) -> str:
    """éè¿´æ›¿æ›å­—ä¸²ä¸­çš„å­—å…ƒ"""
    # åŸºæœ¬æƒ…æ³ï¼šç©ºå­—ä¸²
    if not s:
        return ""
    
    # æª¢æŸ¥ç¬¬ä¸€å€‹å­—å…ƒ
    first_char = new_char if s[0] == old_char else s[0]
    
    # éè¿´è™•ç†å‰©é¤˜å­—ä¸²
    return first_char + replace_char_recursive(s[1:], old_char, new_char)

print("\nä¸­ç­‰é¡Œ 10-12 è§£ç­”å·²å®Œæˆ")
````

````python
# ç¿’é¡Œ 13 è§£ç­”: éè¿´å¯¦ä½œå¿«é€Ÿæ’åº
def quicksort_recursive(arr: list) -> list:
    """éè¿´å¯¦ä½œå¿«é€Ÿæ’åºæ¼”ç®—æ³•"""
    # åŸºæœ¬æƒ…æ³ï¼šé•·åº¦ <= 1
    if len(arr) <= 1:
        return arr
    
    # é¸æ“‡åŸºæº–é»ï¼ˆä¸­é–“å…ƒç´ ï¼‰
    pivot = arr[len(arr) // 2]
    
    # åˆ†å‰²é™£åˆ—
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    
    # éè¿´æ’åºä¸¦åˆä½µ
    return quicksort_recursive(left) + middle + quicksort_recursive(right)

# æ¸¬è©¦
print("ç¿’é¡Œ 13 æ¸¬è©¦ï¼š")
test_arr = [3, 6, 8, 10, 1, 2, 1]
result = quicksort_recursive(test_arr)
print(f"quicksort_recursive({test_arr}) = {result}")
````

````python
# ç¿’é¡Œ 14 è§£ç­”: éè¿´è§£æ±º N çš‡åå•é¡Œ
def n_queens_count(n: int) -> int:
    """éè¿´è¨ˆç®— N çš‡åå•é¡Œçš„è§£æ³•æ•¸é‡"""
    def is_safe(queens: list, row: int, col: int) -> bool:
        """æª¢æŸ¥åœ¨ (row, col) ä½ç½®æ”¾ç½®çš‡åæ˜¯å¦å®‰å…¨"""
        for i, queen_col in enumerate(queens):
            # æª¢æŸ¥åˆ—è¡çª
            if queen_col == col:
                return False
            # æª¢æŸ¥å°è§’ç·šè¡çª
            if abs(i - row) == abs(queen_col - col):
                return False
        return True
    
    def solve(queens: list, row: int) -> int:
        """éè¿´æ±‚è§£ N çš‡åå•é¡Œ"""
        # åŸºæœ¬æƒ…æ³ï¼šæ‰€æœ‰çš‡åéƒ½å·²æ”¾ç½®
        if row == n:
            return 1
        
        count = 0
        # å˜—è©¦åœ¨ç•¶å‰è¡Œçš„æ¯ä¸€åˆ—æ”¾ç½®çš‡å
        for col in range(n):
            if is_safe(queens, row, col):
                queens.append(col)
                count += solve(queens, row + 1)
                queens.pop()  # å›æº¯
        
        return count
    
    return solve([], 0)

# æ¸¬è©¦
print("\nç¿’é¡Œ 14 æ¸¬è©¦ï¼š")
print(f"n_queens_count(4) = {n_queens_count(4)} (æœŸæœ›: 2)")
print(f"n_queens_count(8) = {n_queens_count(8)} (æœŸæœ›: 92)")
````

````python
# ç¿’é¡Œ 15 è§£ç­”: éè¿´è¨ˆç®—æ‰€æœ‰å­é›†
def power_set_recursive(nums: list) -> list:
    """éè¿´è¨ˆç®—é™£åˆ—çš„æ‰€æœ‰å­é›†"""
    # åŸºæœ¬æƒ…æ³ï¼šç©ºé™£åˆ—
    if not nums:
        return [[]]
    
    # å–ç¬¬ä¸€å€‹å…ƒç´ 
    first = nums[0]
    rest = nums[1:]
    
    # éè¿´å–å¾—å‰©é¤˜å…ƒç´ çš„æ‰€æœ‰å­é›†
    subsets_without_first = power_set_recursive(rest)
    
    # åŒ…å«ç¬¬ä¸€å€‹å…ƒç´ çš„å­é›†
    subsets_with_first = [[first] + subset for subset in subsets_without_first]
    
    # åˆä½µå…©ç¨®å­é›†
    return subsets_without_first + subsets_with_first

# ç¿’é¡Œ 16 è§£ç­”: éè¿´å¯¦ä½œè¿·å®®è·¯å¾‘æœå°‹
def maze_paths_count(maze: list, start: tuple, end: tuple) -> int:
    """éè¿´è¨ˆç®—è¿·å®®ä¸­å¾èµ·é»åˆ°çµ‚é»çš„è·¯å¾‘æ•¸é‡"""
    rows, cols = len(maze), len(maze[0])
    visited = [[False for _ in range(cols)] for _ in range(rows)]
    
    def is_valid(row: int, col: int) -> bool:
        """æª¢æŸ¥åº§æ¨™æ˜¯å¦æœ‰æ•ˆä¸”å¯é€šè¡Œ"""
        return (0 <= row < rows and 0 <= col < cols and 
                maze[row][col] == 0 and not visited[row][col])
    
    def find_paths(row: int, col: int) -> int:
        """éè¿´å°‹æ‰¾è·¯å¾‘"""
        # åŸºæœ¬æƒ…æ³ï¼šåˆ°é”çµ‚é»
        if (row, col) == end:
            return 1
        
        # æ¨™è¨˜ç•¶å‰ä½ç½®ç‚ºå·²è¨ªå•
        visited[row][col] = True
        
        # å››å€‹æ–¹å‘
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        total_paths = 0
        
        for dr, dc in directions:
            new_row, new_col = row + dr, col + dc
            if is_valid(new_row, new_col):
                total_paths += find_paths(new_row, new_col)
        
        # å›æº¯ï¼šå–æ¶ˆæ¨™è¨˜
        visited[row][col] = False
        
        return total_paths
    
    return find_paths(start[0], start[1])

print("\né€²éšé¡Œ 15-16 è§£ç­”å·²å®Œæˆ")
````

````python
# ç¿’é¡Œ 17 è§£ç­”: éè¿´å¯¦ä½œè¡¨é”å¼æ±‚å€¼ï¼ˆç°¡åŒ–ç‰ˆï¼‰
def evaluate_expression(expr: str) -> float:
    """éè¿´è¨ˆç®—æ•¸å­¸è¡¨é”å¼çš„å€¼ï¼ˆç°¡åŒ–ç‰ˆï¼‰"""
    expr = expr.replace(' ', '')
    
    def parse_number(s: str, index: int):
        """å¾æŒ‡å®šä½ç½®è§£ææ•¸å­—"""
        start = index
        while index < len(s) and (s[index].isdigit() or s[index] == '.'):
            index += 1
        return float(s[start:index]), index
    
    def parse_expression(s: str, index: int):
        """è§£æåŠ æ¸›è¡¨é”å¼"""
        result, index = parse_term(s, index)
        
        while index < len(s) and s[index] in '+-':
            op = s[index]
            index += 1
            operand, index = parse_term(s, index)
            if op == '+':
                result += operand
            else:
                result -= operand
        
        return result, index
    
    def parse_term(s: str, index: int):
        """è§£æä¹˜é™¤é …"""
        result, index = parse_factor(s, index)
        
        while index < len(s) and s[index] in '*/':
            op = s[index]
            index += 1
            operand, index = parse_factor(s, index)
            if op == '*':
                result *= operand
            else:
                result /= operand
        
        return result, index
    
    def parse_factor(s: str, index: int):
        """è§£æå› å­ï¼ˆæ•¸å­—æˆ–æ‹¬è™Ÿå…§è¡¨é”å¼ï¼‰"""
        if s[index] == '(':
            index += 1  # è·³é '('
            result, index = parse_expression(s, index)
            index += 1  # è·³é ')'
            return result, index
        else:
            return parse_number(s, index)
    
    result, _ = parse_expression(expr, 0)
    return result

# ç¿’é¡Œ 18 è§£ç­”: éè¿´å¯¦ä½œå­—ä¸²åˆ†å‰²çš„æœ€å°æˆæœ¬
def min_word_break_cost(s: str, word_dict: list) -> int:
    """éè¿´è¨ˆç®—å­—ä¸²åˆ†å‰²æˆå­—å…¸å–®è©çš„æœ€å°æˆæœ¬"""
    word_set = set(word_dict)
    memo = {}
    
    def min_cost_helper(start_index: int) -> int:
        """éè¿´è¨ˆç®—å¾ start_index é–‹å§‹çš„æœ€å°æˆæœ¬"""
        # æª¢æŸ¥è¨˜æ†¶åŒ–
        if start_index in memo:
            return memo[start_index]
        
        # åŸºæœ¬æƒ…æ³ï¼šå·²è™•ç†å®Œæ•´å€‹å­—ä¸²
        if start_index == len(s):
            return 0
        
        min_cost = float('inf')
        
        # å˜—è©¦æ¯å€‹å¯èƒ½çš„å‰ç¶´
        for end_index in range(start_index + 1, len(s) + 1):
            prefix = s[start_index:end_index]
            
            # å¦‚æœå‰ç¶´åœ¨å­—å…¸ä¸­
            if prefix in word_set:
                cost = len(prefix) ** 2  # æˆæœ¬ = é•·åº¦çš„å¹³æ–¹
                remaining_cost = min_cost_helper(end_index)
                
                if remaining_cost != float('inf'):
                    min_cost = min(min_cost, cost + remaining_cost)
        
        memo[start_index] = min_cost
        return min_cost
    
    result = min_cost_helper(0)
    return result if result != float('inf') else -1

print("\né€²éšé¡Œ 17-18 è§£ç­”å·²å®Œæˆ")
````
