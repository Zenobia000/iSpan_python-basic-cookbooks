2. **æ­·å²è¨˜éŒ„æ ¼å¼**ï¼š
   - å„²å­˜ç‚ºå­—ä¸²ï¼š`"10 + 5 = 15"`

3. **é™¤æ³•è™•ç†**ï¼š
   - æª¢æŸ¥é™¤æ•¸æ˜¯å¦ç‚ºé›¶
   - éŒ¯èª¤æ™‚å›å‚³è¨Šæ¯ä½†ä¸åŠ å…¥æ­·å²

### å®Œæ•´ç¨‹å¼ç¢¼

```python
class Calculator:
    """
    ç°¡æ˜“è¨ˆç®—æ©Ÿé¡åˆ¥
    
    åŠŸèƒ½ï¼šå››å‰‡é‹ç®— + æ­·å²è¨˜éŒ„
    """
    
    def __init__(self):
        """åˆå§‹åŒ–è¨ˆç®—æ©Ÿï¼Œå‰µå»ºç©ºçš„æ­·å²è¨˜éŒ„"""
        self.history = []  # é‹ç®—æ­·å²åˆ—è¡¨
    
    def add(self, a, b):
        """åŠ æ³•é‹ç®—"""
        result = a + b
        self.history.append(f"{a} + {b} = {result}")
        return result
    
    def subtract(self, a, b):
        """æ¸›æ³•é‹ç®—"""
        result = a - b
        self.history.append(f"{a} - {b} = {result}")
        return result
    
    def multiply(self, a, b):
        """ä¹˜æ³•é‹ç®—"""
        result = a * b
        self.history.append(f"{a} * {b} = {result}")
        return result
    
    def divide(self, a, b):
        """é™¤æ³•é‹ç®—ï¼ˆè™•ç†é™¤ä»¥é›¶ï¼‰"""
        if b == 0:
            return "éŒ¯èª¤ï¼šé™¤æ•¸ä¸èƒ½ç‚ºé›¶"
        result = a / b
        self.history.append(f"{a} / {b} = {result}")
        return result
    
    def show_history(self):
        """é¡¯ç¤ºæ‰€æœ‰é‹ç®—æ­·å²"""
        if not self.history:
            print("å°šç„¡é‹ç®—è¨˜éŒ„")
            return
        
        print("=== é‹ç®—æ­·å² ===")
        for i, record in enumerate(self.history, 1):
            print(f"{i}. {record}")
    
    def clear_history(self):
        """æ¸…ç©ºæ­·å²è¨˜éŒ„"""
        self.history = []
        print("æ­·å²è¨˜éŒ„å·²æ¸…ç©º")

# æ¸¬è©¦è¨ˆç®—æ©Ÿ
calc = Calculator()

# é€²è¡Œå„ç¨®é‹ç®—
print(f"10 + 5 = {calc.add(10, 5)}")
print(f"20 - 8 = {calc.subtract(20, 8)}")
print(f"6 * 7 = {calc.multiply(6, 7)}")
print(f"15 / 3 = {calc.divide(15, 3)}")
print(f"10 / 0 = {calc.divide(10, 0)}")

print()
calc.show_history()

print()
calc.clear_history()
calc.show_history()
```

### å•é¡Œæè¿°
è¨­è¨ˆä¸€å€‹ `Student` é¡åˆ¥ï¼Œç®¡ç†å­¸ç”Ÿè³‡æ–™èˆ‡æˆç¸¾ã€‚

**éœ€æ±‚**ï¼š
1. å„²å­˜å­¸ç”Ÿå§“åã€å­¸è™Ÿã€ç§‘ç›®æˆç¸¾ï¼ˆå­—å…¸ï¼‰
2. æä¾› `add_score()` æ–°å¢æˆç¸¾
3. æä¾› `get_average()` è¨ˆç®—å¹³å‡åˆ†æ•¸
4. æä¾› `get_grade()` å–å¾—ç­‰ç¬¬ï¼ˆA/B/C/D/Fï¼‰
5. æä¾› `show_info()` é¡¯ç¤ºå®Œæ•´è³‡è¨Š

### è§£é¡Œæ€è·¯
1. **å±¬æ€§è¨­è¨ˆ**ï¼š
   - `name`, `student_id`ï¼šåŸºæœ¬è³‡æ–™
   - `scores`ï¼šå­—å…¸å„²å­˜ `{"ç§‘ç›®": åˆ†æ•¸}`

2. **æ–¹æ³•è¨­è¨ˆ**ï¼š
   - å¹³å‡åˆ†æ•¸ï¼š`sum(scores.values()) / len(scores)`
   - ç­‰ç¬¬åˆ¤æ–·ï¼šä¾å¹³å‡åˆ†æ•¸å°æ‡‰ç­‰ç¬¬

### å®Œæ•´ç¨‹å¼ç¢¼

```python
class Student:
    """
    å­¸ç”Ÿæˆç¸¾ç®¡ç†é¡åˆ¥
    
    åŠŸèƒ½ï¼šç®¡ç†å­¸ç”Ÿè³‡æ–™ã€æˆç¸¾ã€è¨ˆç®—å¹³å‡èˆ‡ç­‰ç¬¬
    """
    
    def __init__(self, name, student_id):
        """åˆå§‹åŒ–å­¸ç”Ÿè³‡æ–™"""
        self.name = name
        self.student_id = student_id
        self.scores = {}  # ç§‘ç›®æˆç¸¾å­—å…¸
    
    def add_score(self, subject, score):
        """
        æ–°å¢æˆ–æ›´æ–°ç§‘ç›®æˆç¸¾
        
        åƒæ•¸:
            subject (str): ç§‘ç›®åç¨±
            score (float): åˆ†æ•¸
        """
        if 0 <= score <= 100:
            self.scores[subject] = score
            print(f"âœ“ {self.name} çš„ {subject} æˆç¸¾å·²è¨˜éŒ„ï¼š{score} åˆ†")
        else:
            print(f"âœ— åˆ†æ•¸å¿…é ˆåœ¨ 0-100 ä¹‹é–“")
    
    def get_average(self):
        """è¨ˆç®—å¹³å‡åˆ†æ•¸"""
        if not self.scores:
            return 0
        return sum(self.scores.values()) / len(self.scores)
    
    def get_grade(self):
        """å–å¾—ç­‰ç¬¬ï¼ˆåŸºæ–¼å¹³å‡åˆ†æ•¸ï¼‰"""
        avg = self.get_average()
        if avg >= 90:
            return 'A'
        elif avg >= 80:
            return 'B'
        elif avg >= 70:
            return 'C'
        elif avg >= 60:
            return 'D'
        else:
            return 'F'
    
    def show_info(self):
        """é¡¯ç¤ºå®Œæ•´å­¸ç”Ÿè³‡è¨Š"""
        print(f"\n{'='*40}")
        print(f"å­¸ç”Ÿå§“åï¼š{self.name}")
        print(f"å­¸ã€€ã€€è™Ÿï¼š{self.student_id}")
        print(f"-" * 40)
        
        if self.scores:
            print("ç§‘ç›®æˆç¸¾ï¼š")
            for subject, score in self.scores.items():
                print(f"  {subject:8s}: {score:5.1f} åˆ†")
            print(f"-" * 40)
            print(f"å¹³å‡åˆ†æ•¸ï¼š{self.get_average():.2f} åˆ†")
            print(f"å­¸æœŸç­‰ç¬¬ï¼š{self.get_grade()}")
        else:
            print("å°šç„¡æˆç¸¾è¨˜éŒ„")
        print(f"{'='*40}")

# æ¸¬è©¦å­¸ç”Ÿç®¡ç†
student = Student("ç‹å°æ˜", "A123456789")

# æ–°å¢æˆç¸¾
student.add_score("åœ‹æ–‡", 85)
student.add_score("è‹±æ–‡", 92)
student.add_score("æ•¸å­¸", 88)
student.add_score("ç‰©ç†", 78)
student.add_score("åŒ–å­¸", 150)  # ç„¡æ•ˆåˆ†æ•¸

# é¡¯ç¤ºå®Œæ•´è³‡è¨Š
student.show_info()

# å‰µå»ºå¦ä¸€å€‹å­¸ç”Ÿ
student2 = Student("æå°è¯", "B987654321")
student2.add_score("åœ‹æ–‡", 95)
student2.add_score("è‹±æ–‡", 98)
student2.show_info()
```

## ç¯„ä¾‹ 3ï¼šè³¼ç‰©è»Šç³»çµ±

### å•é¡Œæè¿°
è¨­è¨ˆä¸€å€‹ `ShoppingCart` é¡åˆ¥ï¼Œå¯¦ä½œè³¼ç‰©è»ŠåŠŸèƒ½ã€‚

**éœ€æ±‚**ï¼š
1. å¯æ–°å¢å•†å“ï¼ˆåç¨±ã€åƒ¹æ ¼ã€æ•¸é‡ï¼‰
2. å¯ç§»é™¤å•†å“
3. å¯ä¿®æ”¹å•†å“æ•¸é‡
4. è¨ˆç®—ç¸½é‡‘é¡

### è§£é¡Œæ€è·¯
1. **è³‡æ–™çµæ§‹**ï¼š
   - `items` åˆ—è¡¨å„²å­˜å•†å“å­—å…¸ï¼š`{"name": ..., "price": ..., "quantity": ...}`

2. **æ–¹æ³•è¨­è¨ˆ**ï¼š
   - `add_item()`ï¼šæª¢æŸ¥æ˜¯å¦å·²å­˜åœ¨ï¼Œå­˜åœ¨å‰‡å¢åŠ æ•¸é‡
   - `remove_item()`ï¼šå¾åˆ—è¡¨ä¸­ç§»é™¤
   - `update_quantity()`ï¼šæ›´æ–°æ•¸é‡ï¼Œè‹¥ç‚º 0 å‰‡ç§»é™¤
   - `get_total()`ï¼šè¨ˆç®—æ‰€æœ‰å•†å“é‡‘é¡ç¸½å’Œ

### å®Œæ•´ç¨‹å¼ç¢¼

```python
class ShoppingCart:
    """
    è³¼ç‰©è»Šç³»çµ±
    
    åŠŸèƒ½ï¼šæ–°å¢/ç§»é™¤/ä¿®æ”¹å•†å“ã€è¨ˆç®—ç¸½é‡‘é¡
    """
    
    def __init__(self):
        """åˆå§‹åŒ–ç©ºè³¼ç‰©è»Š"""
        self.items = []  # å•†å“åˆ—è¡¨
    
    def add_item(self, name, price, quantity=1):
        """
        æ–°å¢å•†å“åˆ°è³¼ç‰©è»Š
        
        åƒæ•¸:
            name (str): å•†å“åç¨±
            price (float): å–®åƒ¹
            quantity (int): æ•¸é‡ï¼ˆé è¨­ 1ï¼‰
        """
        # æª¢æŸ¥å•†å“æ˜¯å¦å·²å­˜åœ¨
        for item in self.items:
            if item["name"] == name:
                item["quantity"] += quantity
                print(f"âœ“ {name} æ•¸é‡å·²å¢åŠ  {quantity}ï¼Œç›®å‰å…± {item['quantity']} ä»¶")
                return
        
        # æ–°å•†å“
        self.items.append({
            "name": name,
            "price": price,
            "quantity": quantity
        })
        print(f"âœ“ {name} å·²åŠ å…¥è³¼ç‰©è»Šï¼ˆ{quantity} ä»¶ï¼‰")
    
    def remove_item(self, name):
        """ç§»é™¤å•†å“"""
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                print(f"âœ“ {name} å·²å¾è³¼ç‰©è»Šç§»é™¤")
                return
        print(f"âœ— è³¼ç‰©è»Šä¸­æ²’æœ‰ {name}")
    
    def update_quantity(self, name, quantity):
        """æ›´æ–°å•†å“æ•¸é‡"""
        for item in self.items:
            if item["name"] == name:
                if quantity <= 0:
                    self.remove_item(name)
                else:
                    item["quantity"] = quantity
                    print(f"âœ“ {name} æ•¸é‡å·²æ›´æ–°ç‚º {quantity} ä»¶")
                return
        print(f"âœ— è³¼ç‰©è»Šä¸­æ²’æœ‰ {name}")
    
    def get_total(self):
        """è¨ˆç®—ç¸½é‡‘é¡"""
        return sum(item["price"] * item["quantity"] for item in self.items)
    
    def show_cart(self):
        """é¡¯ç¤ºè³¼ç‰©è»Šå…§å®¹"""
        if not self.items:
            print("è³¼ç‰©è»Šæ˜¯ç©ºçš„")
            return
        
        print("\n" + "="*60)
        print(f"{'å•†å“åç¨±':20s} {'å–®åƒ¹':>10s} {'æ•¸é‡':>8s} {'å°è¨ˆ':>15s}")
        print("-" * 60)
        
        for item in self.items:
            subtotal = item["price"] * item["quantity"]
            print(f"{item['name']:20s} ${item['price']:>9.2f} {item['quantity']:>8d} ${subtotal:>14.2f}")
        
        print("-" * 60)
        print(f"{'ç¸½è¨ˆ':>50s} ${self.get_total():>7.2f}")
        print("=" * 60)

# æ¸¬è©¦è³¼ç‰©è»Š
cart = ShoppingCart()

# æ–°å¢å•†å“
cart.add_item("è˜‹æœ", 30, 3)
cart.add_item("ç‰›å¥¶", 50, 2)
cart.add_item("éºµåŒ…", 40, 1)
cart.add_item("è˜‹æœ", 30, 2)  # å¢åŠ ç¾æœ‰å•†å“æ•¸é‡

# é¡¯ç¤ºè³¼ç‰©è»Š
cart.show_cart()

# ä¿®æ”¹æ•¸é‡
print()
cart.update_quantity("ç‰›å¥¶", 5)

# ç§»é™¤å•†å“
cart.remove_item("éºµåŒ…")

# å†æ¬¡é¡¯ç¤º
cart.show_cart()
```

1. å¯å„²å­˜æ”æ°æˆ–è¯æ°æº«åº¦

### è§£é¡Œæ€è·¯
1. **æ–¹æ³•é¡å‹é¸æ“‡**ï¼š
   - å¯¦ä¾‹æ–¹æ³•ï¼šæ“ä½œå¯¦ä¾‹è³‡æ–™
   - é¡åˆ¥æ–¹æ³•ï¼šä½œç‚ºæ›¿ä»£å»ºæ§‹å­
   - éœæ…‹æ–¹æ³•ï¼šç¨ç«‹çš„å·¥å…·å‡½å¼

2. **è¨­è¨ˆæ¨¡å¼**ï¼š
   - å…§éƒ¨çµ±ä¸€å„²å­˜æ”æ°æº«åº¦
   - å‹•æ…‹è½‰æ›ç‚ºè¯æ°æº«åº¦

### å®Œæ•´ç¨‹å¼ç¢¼

```python
class Temperature:
    """
    æº«åº¦è½‰æ›å™¨é¡åˆ¥
    
    åŠŸèƒ½ï¼šæ”æ°è¯æ°è½‰æ›ã€æº«åº¦é©—è­‰
    """
    
    def __init__(self, celsius):
        """åˆå§‹åŒ–æº«åº¦ï¼ˆæ”æ°ï¼‰"""
        if not Temperature.is_valid_celsius(celsius):
            raise ValueError("æº«åº¦ä¸èƒ½ä½æ–¼çµ•å°é›¶åº¦ï¼ˆ-273.15Â°Cï¼‰")
        self.celsius = celsius
    
    @classmethod
    def from_fahrenheit(cls, fahrenheit):
        """
        å¾è¯æ°æº«åº¦å‰µå»ºå¯¦ä¾‹ï¼ˆé¡åˆ¥æ–¹æ³•ï¼‰
        
        åƒæ•¸:
            fahrenheit (float): è¯æ°æº«åº¦
        
        å›å‚³:
            Temperature: æº«åº¦ç‰©ä»¶
        """
        celsius = (fahrenheit - 32) * 5 / 9
        return cls(celsius)
    
    @staticmethod
    def is_valid_celsius(celsius):
        """
        é©—è­‰æ”æ°æº«åº¦æ˜¯å¦æœ‰æ•ˆï¼ˆéœæ…‹æ–¹æ³•ï¼‰
        
        åƒæ•¸:
            celsius (float): æ”æ°æº«åº¦
        
        å›å‚³:
            bool: æ˜¯å¦æœ‰æ•ˆ
        """
        return celsius >= -273.15  # çµ•å°é›¶åº¦
    
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        """æ”æ°è½‰è¯æ°ï¼ˆéœæ…‹æ–¹æ³•ï¼‰"""
        return celsius * 9 / 5 + 32
    
    @staticmethod
    def fahrenheit_to_celsius(fahrenheit):
        """è¯æ°è½‰æ”æ°ï¼ˆéœæ…‹æ–¹æ³•ï¼‰"""
        return (fahrenheit - 32) * 5 / 9
    
    def to_fahrenheit(self):
        """å–å¾—è¯æ°æº«åº¦ï¼ˆå¯¦ä¾‹æ–¹æ³•ï¼‰"""
        return Temperature.celsius_to_fahrenheit(self.celsius)
    
    def show_both(self):
        """é¡¯ç¤ºå…©ç¨®æº«åº¦ï¼ˆå¯¦ä¾‹æ–¹æ³•ï¼‰"""
        fahrenheit = self.to_fahrenheit()
        print(f"æ”æ°ï¼š{self.celsius:.2f}Â°C")
        print(f"è¯æ°ï¼š{fahrenheit:.2f}Â°F")

# æ¸¬è©¦æº«åº¦è½‰æ›å™¨
print("=== å¾æ”æ°å‰µå»º ===")
temp1 = Temperature(25)
temp1.show_both()

print("\n=== å¾è¯æ°å‰µå»ºï¼ˆé¡åˆ¥æ–¹æ³•ï¼‰===")
temp2 = Temperature.from_fahrenheit(98.6)
temp2.show_both()

print("\n=== éœæ…‹æ–¹æ³•æ¸¬è©¦ ===")
print(f"100Â°C = {Temperature.celsius_to_fahrenheit(100)}Â°F")
print(f"32Â°F = {Temperature.fahrenheit_to_celsius(32)}Â°C")
print(f"-300Â°C æœ‰æ•ˆå—ï¼Ÿ{Temperature.is_valid_celsius(-300)}")
print(f"0Â°C æœ‰æ•ˆå—ï¼Ÿ{Temperature.is_valid_celsius(0)}")

print("\n=== éŒ¯èª¤è™•ç† ===")
try:
    invalid_temp = Temperature(-300)
except ValueError as e:
    print(f"éŒ¯èª¤ï¼š{e}")
```

```python
class GameCharacter:
    """
    éŠæˆ²è§’è‰²é¡åˆ¥
    
    åŠŸèƒ½ï¼šè§’è‰²å±¬æ€§ã€æˆ°é¬¥ã€æ²»ç™‚ã€å‡ç´š
    """
    
    total_characters = 0  # é¡åˆ¥å±¬æ€§ï¼šè¿½è¹¤ç¸½è§’è‰²æ•¸
    
    def __init__(self, name, hp=100, attack=10, level=1):
        """åˆå§‹åŒ–è§’è‰²"""
        self.name = name
        self.hp = hp
        self.max_hp = hp
        self.attack = attack
        self.level = level
        
        # å¢åŠ ç¸½è§’è‰²æ•¸
        GameCharacter.total_characters += 1
        print(f"âœ“ è§’è‰² {self.name} å·²å‰µå»ºï¼ˆç­‰ç´š {self.level}ï¼‰")
    
    def attack_target(self, target):
        """
        æ”»æ“Šç›®æ¨™è§’è‰²
        
        åƒæ•¸:
            target (GameCharacter): ç›®æ¨™è§’è‰²
        """
        if self.hp <= 0:
            print(f"{self.name} å·²é™£äº¡ï¼Œç„¡æ³•æ”»æ“Š")
            return
        
        damage = self.attack
        target.take_damage(damage)
        print(f"âš”ï¸ {self.name} æ”»æ“Š {target.name}ï¼Œé€ æˆ {damage} é»å‚·å®³")
    
    def take_damage(self, damage):
        """å—åˆ°å‚·å®³"""
        self.hp -= damage
        if self.hp < 0:
            self.hp = 0
        
        if self.hp == 0:
            print(f"ğŸ’€ {self.name} å·²é™£äº¡")
    
    def heal(self, amount):
        """æ²»ç™‚æ¢å¾© HP"""
        if self.hp == 0:
            print(f"{self.name} å·²é™£äº¡ï¼Œç„¡æ³•æ²»ç™‚")
            return
        
        old_hp = self.hp
        self.hp = min(self.hp + amount, self.max_hp)
        actual_heal = self.hp - old_hp
        print(f"ğŸ’š {self.name} æ¢å¾© {actual_heal} HPï¼ˆ{old_hp} â†’ {self.hp}ï¼‰")
    
    def level_up(self):
        """å‡ç´šï¼ˆæå‡å±¬æ€§ï¼‰"""
        self.level += 1
        self.max_hp += 20
        self.hp = self.max_hp  # å‡ç´šæ™‚æ»¿è¡€
        self.attack += 5
        print(f"â¬†ï¸ {self.name} å‡ç´šåˆ° Lv.{self.level}ï¼")
        print(f"   HP: {self.max_hp}, æ”»æ“ŠåŠ›: {self.attack}")
    
    def show_status(self):
        """é¡¯ç¤ºè§’è‰²ç‹€æ…‹"""
        status = "å­˜æ´»" if self.hp > 0 else "é™£äº¡"
        print(f"\n{'='*40}")
        print(f"è§’è‰²ï¼š{self.name} (Lv.{self.level})")
        print(f"HPï¼š{self.hp}/{self.max_hp}")
        print(f"æ”»æ“ŠåŠ›ï¼š{self.attack}")
        print(f"ç‹€æ…‹ï¼š{status}")
        print(f"{'='*40}")

# æ¸¬è©¦éŠæˆ²è§’è‰²
print("=== å‰µå»ºè§’è‰² ===")
hero = GameCharacter("å‹‡è€…", hp=150, attack=20, level=5)
monster = GameCharacter("é­”ç‹", hp=200, attack=25, level=10)

print(f"\nç›®å‰ç¸½è§’è‰²æ•¸ï¼š{GameCharacter.total_characters}")

print("\n=== åˆå§‹ç‹€æ…‹ ===")
hero.show_status()
monster.show_status()

print("\n=== æˆ°é¬¥é–‹å§‹ ===")
hero.attack_target(monster)
monster.show_status()

monster.attack_target(hero)
hero.show_status()

print("\n=== æ²»ç™‚ ===")
hero.heal(30)

print("\n=== å‡ç´š ===")
hero.level_up()
hero.show_status()
```
