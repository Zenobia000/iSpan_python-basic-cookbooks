### ç·´ç¿’ 1ï¼šRectangle çŸ©å½¢é¡åˆ¥ - è§£ç­”

```python
class Rectangle:
    """çŸ©å½¢é¡åˆ¥"""
    
    def __init__(self, length, width):
        """åˆå§‹åŒ–çŸ©å½¢"""
        self.length = length
        self.width = width
    
    def area(self):
        """è¨ˆç®—é¢ç©"""
        return self.length * self.width
    
    def perimeter(self):
        """è¨ˆç®—å‘¨é•·"""
        return 2 * (self.length + self.width)
    
    def is_square(self):
        """åˆ¤æ–·æ˜¯å¦ç‚ºæ­£æ–¹å½¢"""
        return self.length == self.width

# æ¸¬è©¦
rect1 = Rectangle(10, 5)
print(f"é¢ç©ï¼š{rect1.area()}")        # 50
print(f"å‘¨é•·ï¼š{rect1.perimeter()}")   # 30
print(f"æ˜¯æ­£æ–¹å½¢ï¼š{rect1.is_square()}")   # False

rect2 = Rectangle(8, 8)
print(f"æ˜¯æ­£æ–¹å½¢ï¼š{rect2.is_square()}")   # True
```

### ç·´ç¿’ 2ï¼šCircle åœ“å½¢é¡åˆ¥ - è§£ç­”

```python
class Circle:
    """åœ“å½¢é¡åˆ¥"""
    
    PI = 3.14159  # é¡åˆ¥å±¬æ€§
    
    def __init__(self, radius):
        """åˆå§‹åŒ–åœ“å½¢"""
        self.radius = radius
    
    def area(self):
        """è¨ˆç®—é¢ç©"""
        return Circle.PI * self.radius ** 2
    
    def circumference(self):
        """è¨ˆç®—å‘¨é•·"""
        return 2 * Circle.PI * self.radius
    
    def diameter(self):
        """è¨ˆç®—ç›´å¾‘"""
        return 2 * self.radius

# æ¸¬è©¦
circle = Circle(5)
print(f"é¢ç©ï¼š{circle.area():.2f}")           # 78.54
print(f"å‘¨é•·ï¼š{circle.circumference():.2f}")  # 31.42
print(f"ç›´å¾‘ï¼š{circle.diameter()}")           # 10
```

### ç·´ç¿’ 5ï¼šStudent å­¸ç”Ÿé¡åˆ¥ - è§£ç­”

```python
class Student:
    """å­¸ç”Ÿé¡åˆ¥"""
    
    def __init__(self, name, age):
        """åˆå§‹åŒ–å­¸ç”Ÿ"""
        self.name = name
        self.age = age
        self.grades = []  # æˆç¸¾åˆ—è¡¨
    
    def add_grade(self, score):
        """æ–°å¢æˆç¸¾"""
        self.grades.append(score)
    
    def get_average(self):
        """è¨ˆç®—å¹³å‡åˆ†æ•¸"""
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)
    
    def show_report(self):
        """é¡¯ç¤ºå­¸ç”Ÿå ±å‘Š"""
        print(f"å­¸ç”Ÿï¼š{self.name}ï¼ˆ{self.age} æ­²ï¼‰")
        print(f"æˆç¸¾ï¼š{self.grades}")
        print(f"å¹³å‡ï¼š{self.get_average():.2f} åˆ†")

# æ¸¬è©¦
student = Student("ç‹å°æ˜", 18)
student.add_grade(85)
student.add_grade(90)
student.add_grade(88)
print(f"å¹³å‡ï¼š{student.get_average():.2f}")  # 87.67
student.show_report()
```

### ç·´ç¿’ 6ï¼šProduct å•†å“é¡åˆ¥ - è§£ç­”

```python
class Product:
    """å•†å“é¡åˆ¥"""
    
    def __init__(self, name, price, quantity):
        """åˆå§‹åŒ–å•†å“"""
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def sell(self, amount):
        """éŠ·å”®å•†å“"""
        if amount > self.quantity:
            print(f"éŒ¯èª¤ï¼šåº«å­˜ä¸è¶³ï¼ˆå‰©é¤˜ {self.quantity}ï¼‰")
            return
        self.quantity -= amount
        print(f"å·²å”®å‡º {amount} ä»¶ {self.name}")
    
    def restock(self, amount):
        """è£œè²¨"""
        self.quantity += amount
        print(f"å·²è£œè²¨ {amount} ä»¶")
    
    def get_value(self):
        """è¨ˆç®—ç¸½åƒ¹å€¼"""
        return self.price * self.quantity

# æ¸¬è©¦
product = Product("ç­†è¨˜å‹é›»è…¦", 30000, 10)
product.sell(3)
print(f"å‰©é¤˜åº«å­˜ï¼š{product.quantity}")  # 7
print(f"åº«å­˜åƒ¹å€¼ï¼š{product.get_value()}")  # 210000

product.restock(5)
print(f"å‰©é¤˜åº«å­˜ï¼š{product.quantity}")  # 12
```

### ç·´ç¿’ 7ï¼šTemperature æº«åº¦é¡åˆ¥ - è§£ç­”

```python
class Temperature:
    """æº«åº¦é¡åˆ¥"""
    
    def __init__(self, celsius):
        """åˆå§‹åŒ–æº«åº¦ï¼ˆæ”æ°ï¼‰"""
        self.celsius = celsius
    
    def to_fahrenheit(self):
        """è½‰æ›ç‚ºè¯æ°æº«åº¦"""
        return self.celsius * 9/5 + 32
    
    def to_kelvin(self):
        """è½‰æ›ç‚ºçµ•å°æº«åº¦"""
        return self.celsius + 273.15
    
    def is_freezing(self):
        """åˆ¤æ–·æ˜¯å¦ä½æ–¼ 0Â°C"""
        return self.celsius < 0
    
    def is_boiling(self):
        """åˆ¤æ–·æ˜¯å¦é«˜æ–¼ 100Â°C"""
        return self.celsius > 100

# æ¸¬è©¦
temp = Temperature(25)
print(f"è¯æ°ï¼š{temp.to_fahrenheit()}")  # 77.0
print(f"çµ•å°æº«åº¦ï¼š{temp.to_kelvin()}")  # 298.15
print(f"çµå†°ï¼š{temp.is_freezing()}")  # False
print(f"æ²¸é¨°ï¼š{temp.is_boiling()}")   # False

temp2 = Temperature(-5)
print(f"çµå†°ï¼š{temp2.is_freezing()}")  # True
```

### ç·´ç¿’ 8ï¼šTimer è¨ˆæ™‚å™¨é¡åˆ¥ - è§£ç­”

```python
class Timer:
    """è¨ˆæ™‚å™¨é¡åˆ¥"""
    
    def __init__(self):
        """åˆå§‹åŒ–è¨ˆæ™‚å™¨"""
        self.seconds = 0
    
    def set_time(self, hours, minutes, seconds):
        """è¨­å®šæ™‚é–“ï¼ˆè½‰æ›ç‚ºç¸½ç§’æ•¸ï¼‰"""
        self.seconds = hours * 3600 + minutes * 60 + seconds
    
    def add_seconds(self, sec):
        """å¢åŠ ç§’æ•¸"""
        self.seconds += sec
    
    def get_time(self):
        """å›å‚³æ ¼å¼åŒ–æ™‚é–“ï¼ˆHH:MM:SSï¼‰"""
        hours = self.seconds // 3600
        minutes = (self.seconds % 3600) // 60
        seconds = self.seconds % 60
        return f"{hours:02d}:{minutes:02d}:{seconds:02d}"

# æ¸¬è©¦
timer = Timer()
timer.set_time(1, 30, 45)
print(timer.get_time())  # "01:30:45"

timer.add_seconds(20)
print(timer.get_time())  # "01:31:05"

timer.add_seconds(100)
print(timer.get_time())  # "01:32:45"
```

### ç·´ç¿’ 9ï¼šBankAccount éŠ€è¡Œå¸³æˆ¶ - è§£ç­”

```python
class BankAccount:
    """éŠ€è¡Œå¸³æˆ¶é¡åˆ¥"""
    
    interest_rate = 0.02  # é¡åˆ¥å±¬æ€§ï¼šå¹´åˆ©ç‡
    total_accounts = 0    # é¡åˆ¥å±¬æ€§ï¼šç¸½å¸³æˆ¶æ•¸
    
    def __init__(self, owner, balance=0):
        """åˆå§‹åŒ–å¸³æˆ¶"""
        self.owner = owner
        self.balance = balance
        BankAccount.total_accounts += 1
    
    def deposit(self, amount):
        """å­˜æ¬¾"""
        self.balance += amount
        print(f"{self.owner} å­˜æ¬¾ ${amount}ï¼Œé¤˜é¡ ${self.balance}")
    
    def withdraw(self, amount):
        """ææ¬¾"""
        if amount > self.balance:
            print(f"éŒ¯èª¤ï¼šé¤˜é¡ä¸è¶³ï¼ˆå‰©é¤˜ ${self.balance}ï¼‰")
            return
        self.balance -= amount
        print(f"{self.owner} ææ¬¾ ${amount}ï¼Œé¤˜é¡ ${self.balance}")
    
    def apply_interest(self):
        """å¥—ç”¨åˆ©æ¯"""
        interest = self.balance * BankAccount.interest_rate
        self.balance += interest
        print(f"åˆ©æ¯ ${interest:.2f} å·²å…¥å¸³")
    
    @classmethod
    def set_interest_rate(cls, rate):
        """è¨­å®šåˆ©ç‡ï¼ˆé¡åˆ¥æ–¹æ³•ï¼‰"""
        cls.interest_rate = rate
        print(f"åˆ©ç‡å·²èª¿æ•´ç‚º {rate*100}%")

# æ¸¬è©¦
account1 = BankAccount("Alice", 1000)
account2 = BankAccount("Bob", 2000)
print(f"ç¸½å¸³æˆ¶æ•¸ï¼š{BankAccount.total_accounts}")  # 2

account1.apply_interest()
print(f"Alice é¤˜é¡ï¼š{account1.balance}")  # 1020.0

BankAccount.set_interest_rate(0.03)
account2.apply_interest()
print(f"Bob é¤˜é¡ï¼š{account2.balance}")  # 2060.0
```

**çŸ¥è­˜é»**ï¼š
- é¡åˆ¥å±¬æ€§ï¼š`interest_rate` æ‰€æœ‰å¸³æˆ¶å…±äº«
- é¡åˆ¥æ–¹æ³•ï¼š`@classmethod` ä¿®æ”¹é¡åˆ¥å±¬æ€§
- é è¨­åƒæ•¸ï¼š`balance=0`
- å»ºæ§‹å­è¨ˆæ•¸ï¼š`total_accounts += 1`

### ç·´ç¿’ 10ï¼šLibraryMember åœ–æ›¸é¤¨æœƒå“¡ - è§£ç­”

```python
class LibraryMember:
    """åœ–æ›¸é¤¨æœƒå“¡é¡åˆ¥"""
    
    total_members = 0     # é¡åˆ¥å±¬æ€§ï¼šç¸½æœƒå“¡æ•¸
    borrowed_books = 0    # é¡åˆ¥å±¬æ€§ï¼šç¸½å€Ÿé–±æ•¸
    
    def __init__(self, name, member_id):
        """åˆå§‹åŒ–æœƒå“¡"""
        self.name = name
        self.member_id = member_id
        self.books_borrowed = 0  # å¯¦ä¾‹å±¬æ€§ï¼šå€‹äººå€Ÿé–±æ•¸
        LibraryMember.total_members += 1
    
    def borrow_book(self):
        """å€Ÿæ›¸"""
        self.books_borrowed += 1
        LibraryMember.borrowed_books += 1
        print(f"{self.name} å€Ÿäº† 1 æœ¬æ›¸ï¼ˆå€‹äººå…± {self.books_borrowed} æœ¬ï¼‰")
    
    def return_book(self):
        """é‚„æ›¸"""
        if self.books_borrowed > 0:
            self.books_borrowed -= 1
            print(f"{self.name} é‚„äº† 1 æœ¬æ›¸ï¼ˆå€‹äººå‰© {self.books_borrowed} æœ¬ï¼‰")
        else:
            print(f"{self.name} æ²’æœ‰å€Ÿæ›¸")
    
    @classmethod
    def get_statistics(cls):
        """é¡¯ç¤ºçµ±è¨ˆè³‡è¨Šï¼ˆé¡åˆ¥æ–¹æ³•ï¼‰"""
        print(f"ç¸½æœƒå“¡æ•¸ï¼š{cls.total_members}")
        print(f"ç¸½å€Ÿé–±æ•¸ï¼š{cls.borrowed_books}")

# æ¸¬è©¦
member1 = LibraryMember("Alice", "M001")
member2 = LibraryMember("Bob", "M002")

member1.borrow_book()
member1.borrow_book()
member2.borrow_book()

LibraryMember.get_statistics()
```

### ç·´ç¿’ 12ï¼šPoint é»åº§æ¨™ - è§£ç­”

```python
import math

class Point:
    """é»åº§æ¨™é¡åˆ¥"""
    
    def __init__(self, x, y):
        """åˆå§‹åŒ–é»"""
        self.x = x
        self.y = y
    
    def distance_to(self, other_point):
        """è¨ˆç®—åˆ°å¦ä¸€é»çš„è·é›¢ï¼ˆå¯¦ä¾‹æ–¹æ³•ï¼‰"""
        dx = self.x - other_point.x
        dy = self.y - other_point.y
        return math.sqrt(dx**2 + dy**2)
    
    @staticmethod
    def distance_between(p1, p2):
        """è¨ˆç®—å…©é»è·é›¢ï¼ˆéœæ…‹æ–¹æ³•ï¼‰"""
        dx = p1.x - p2.x
        dy = p1.y - p2.y
        return math.sqrt(dx**2 + dy**2)
    
    def move(self, dx, dy):
        """ç§»å‹•é»"""
        self.x += dx
        self.y += dy

# æ¸¬è©¦
p1 = Point(0, 0)
p2 = Point(3, 4)

print(f"è·é›¢ï¼ˆå¯¦ä¾‹æ–¹æ³•ï¼‰ï¼š{p1.distance_to(p2)}")  # 5.0
print(f"è·é›¢ï¼ˆéœæ…‹æ–¹æ³•ï¼‰ï¼š{Point.distance_between(p1, p2)}")  # 5.0

p1.move(1, 1)
print(f"æ–°åº§æ¨™ï¼š({p1.x}, {p1.y})")  # (1, 1)
print(f"æ–°è·é›¢ï¼š{p1.distance_to(p2):.2f}")  # 3.61
```

**çŸ¥è­˜é»**ï¼š
- éœæ…‹æ–¹æ³•ï¼š`@staticmethod` ä¸ä¾è³´å¯¦ä¾‹
- å¯¦ä¾‹æ–¹æ³• vs éœæ…‹æ–¹æ³•ï¼šå…©ç¨®è·é›¢è¨ˆç®—æ–¹å¼

### ç·´ç¿’ 13ï¼šTime æ™‚é–“é¡åˆ¥ - è§£ç­”

```python
class Time:
    """æ™‚é–“é¡åˆ¥"""
    
    def __init__(self, hour, minute, second):
        """åˆå§‹åŒ–æ™‚é–“ï¼ˆ24 å°æ™‚åˆ¶ï¼‰"""
        self.hour = hour
        self.minute = minute
        self.second = second
    
    def get_24h_format(self):
        """å›å‚³ 24 å°æ™‚åˆ¶å­—ä¸²"""
        return f"{self.hour:02d}:{self.minute:02d}:{self.second:02d}"
    
    def get_12h_format(self):
        """å›å‚³ 12 å°æ™‚åˆ¶å­—ä¸²"""
        period = "AM" if self.hour < 12 else "PM"
        hour_12 = self.hour if self.hour <= 12 else self.hour - 12
        if hour_12 == 0:
            hour_12 = 12
        return f"{hour_12:02d}:{self.minute:02d}:{self.second:02d} {period}"
    
    @classmethod
    def from_string(cls, time_str):
        """å¾å­—ä¸²å‰µå»ºæ™‚é–“ï¼ˆé¡åˆ¥æ–¹æ³•ï¼‰"""
        parts = time_str.split(":")
        hour = int(parts[0])
        minute = int(parts[1])
        second = int(parts[2])
        return cls(hour, minute, second)
    
    def add_hours(self, h):
        """å¢åŠ å°æ™‚æ•¸ï¼ˆè™•ç†è·¨æ—¥ï¼‰"""
        self.hour = (self.hour + h) % 24

# æ¸¬è©¦
time1 = Time(14, 30, 0)
print(time1.get_24h_format())  # "14:30:00"
print(time1.get_12h_format())  # "02:30:00 PM"

time2 = Time.from_string("09:15:30")
print(time2.get_12h_format())  # "09:15:30 AM"

time1.add_hours(10)
print(time1.get_24h_format())  # "00:30:00"
```

### ç·´ç¿’ 15ï¼šLibrary åœ–æ›¸é¤¨ç®¡ç†ç³»çµ± - è§£ç­”

```python
class Book:
    """æ›¸ç±é¡åˆ¥"""
    
    def __init__(self, title, author, isbn):
        """åˆå§‹åŒ–æ›¸ç±"""
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False


class Library:
    """åœ–æ›¸é¤¨é¡åˆ¥"""
    
    def __init__(self):
        """åˆå§‹åŒ–åœ–æ›¸é¤¨"""
        self.books = []
    
    def add_book(self, book):
        """æ–°å¢æ›¸ç±"""
        self.books.append(book)
        print(f"å·²æ–°å¢æ›¸ç±ï¼š{book.title}")
    
    def borrow_book(self, isbn):
        """å€Ÿæ›¸"""
        for book in self.books:
            if book.isbn == isbn:
                if book.is_borrowed:
                    print(f"ã€Š{book.title}ã€‹å·²è¢«å€Ÿå‡º")
                else:
                    book.is_borrowed = True
                    print(f"æˆåŠŸå€Ÿé–±ï¼šã€Š{book.title}ã€‹")
                return
        print(f"æ‰¾ä¸åˆ° ISBN: {isbn}")
    
    def return_book(self, isbn):
        """é‚„æ›¸"""
        for book in self.books:
            if book.isbn == isbn:
                if book.is_borrowed:
                    book.is_borrowed = False
                    print(f"å·²æ­¸é‚„ï¼šã€Š{book.title}ã€‹")
                else:
                    print(f"ã€Š{book.title}ã€‹æœªè¢«å€Ÿå‡º")
                return
        print(f"æ‰¾ä¸åˆ° ISBN: {isbn}")
    
    def find_book(self, title):
        """æŸ¥è©¢æ›¸ç±ï¼ˆéƒ¨åˆ†æ¯”å°ï¼‰"""
        results = [book for book in self.books if title in book.title]
        if results:
            print(f"æ‰¾åˆ° {len(results)} æœ¬æ›¸ç±ï¼š")
            for book in results:
                status = "å·²å€Ÿå‡º" if book.is_borrowed else "å¯å€Ÿé–±"
                print(f"  - ã€Š{book.title}ã€‹ï¼ˆ{book.author}ï¼‰[{status}]")
        else:
            print(f"æ‰¾ä¸åˆ°åŒ…å« '{title}' çš„æ›¸ç±")
        return results
    
    def list_available_books(self):
        """åˆ—å‡ºå¯å€Ÿæ›¸ç±"""
        available = [book for book in self.books if not book.is_borrowed]
        print(f"å¯å€Ÿé–±æ›¸ç±ï¼ˆå…± {len(available)} æœ¬ï¼‰ï¼š")
        for book in available:
            print(f"  - ã€Š{book.title}ã€‹ï¼ˆ{book.author}ï¼‰- {book.isbn}")
    
    def get_statistics(self):
        """é¡¯ç¤ºçµ±è¨ˆ"""
        total = len(self.books)
        borrowed = sum(1 for book in self.books if book.is_borrowed)
        available = total - borrowed
        print(f"ç¸½æ›¸æ•¸ï¼š{total}")
        print(f"å·²å€Ÿå‡ºï¼š{borrowed}")
        print(f"å¯å€Ÿé–±ï¼š{available}")


# æ¸¬è©¦
library = Library()
library.add_book(Book("Python åŸºç¤", "å¼µä¸‰", "ISBN001"))
library.add_book(Book("Java å…¥é–€", "æå››", "ISBN002"))
library.add_book(Book("Python é€²éš", "ç‹äº”", "ISBN003"))

print()
library.borrow_book("ISBN001")

print()
library.list_available_books()

print()
library.find_book("Python")

print()
library.get_statistics()
```

### ç·´ç¿’ 16ï¼šInventory åº«å­˜ç®¡ç†ç³»çµ± - è§£ç­”

```python
class Product:
    """å•†å“é¡åˆ¥"""
    
    def __init__(self, product_id, name, price, quantity):
        """åˆå§‹åŒ–å•†å“"""
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def get_value(self):
        """è¨ˆç®—ç¸½åƒ¹å€¼"""
        return self.price * self.quantity
    
    def update_quantity(self, delta):
        """èª¿æ•´åº«å­˜"""
        self.quantity += delta
        if self.quantity < 0:
            self.quantity = 0


class Inventory:
    """åº«å­˜ç®¡ç†é¡åˆ¥"""
    
    def __init__(self):
        """åˆå§‹åŒ–åº«å­˜"""
        self.products = {}  # product_id â†’ Product
    
    def add_product(self, product):
        """æ–°å¢å•†å“"""
        self.products[product.product_id] = product
        print(f"å·²æ–°å¢å•†å“ï¼š{product.name}")
    
    def remove_product(self, product_id):
        """ç§»é™¤å•†å“"""
        if product_id in self.products:
            name = self.products[product_id].name
            del self.products[product_id]
            print(f"å·²ç§»é™¤å•†å“ï¼š{name}")
        else:
            print(f"æ‰¾ä¸åˆ°å•†å“ ID: {product_id}")
    
    def update_stock(self, product_id, delta):
        """èª¿æ•´åº«å­˜"""
        if product_id in self.products:
            product = self.products[product_id]
            product.update_quantity(delta)
            action = "å¢åŠ " if delta > 0 else "æ¸›å°‘"
            print(f"{product.name} åº«å­˜{action} {abs(delta)}ï¼Œç›®å‰ {product.quantity}")
        else:
            print(f"æ‰¾ä¸åˆ°å•†å“ ID: {product_id}")
    
    def get_low_stock(self, threshold):
        """å–å¾—ä½åº«å­˜å•†å“"""
        low_stock = [p for p in self.products.values() if p.quantity < threshold]
        print(f"ä½åº«å­˜å•†å“ï¼ˆ< {threshold}ï¼‰ï¼š")
        for product in low_stock:
            print(f"  - {product.name}: {product.quantity}")
        return low_stock
    
    def get_total_value(self):
        """è¨ˆç®—ç¸½åº«å­˜åƒ¹å€¼"""
        return sum(p.get_value() for p in self.products.values())
    
    def generate_report(self):
        """ç”Ÿæˆåº«å­˜å ±å‘Š"""
        print("\n" + "="*60)
        print(f"{'å•†å“åç¨±':20s} {'å–®åƒ¹':>10s} {'åº«å­˜':>8s} {'ç¸½åƒ¹å€¼':>15s}")
        print("-" * 60)
        
        for product in self.products.values():
            value = product.get_value()
            print(f"{product.name:20s} ${product.price:>9.2f} {product.quantity:>8d} ${value:>14.2f}")
        
        print("-" * 60)
        print(f"{'ç¸½åº«å­˜åƒ¹å€¼':>50s} ${self.get_total_value():>7.2f}")
        print("=" * 60)


# æ¸¬è©¦
inventory = Inventory()
inventory.add_product(Product("P001", "ç­†è¨˜å‹é›»è…¦", 30000, 5))
inventory.add_product(Product("P002", "æ»‘é¼ ", 500, 2))
inventory.add_product(Product("P003", "éµç›¤", 1500, 8))

print()
inventory.update_stock("P001", -2)

print()
inventory.get_low_stock(5)

print()
print(f"ç¸½åƒ¹å€¼ï¼š${inventory.get_total_value()}")

inventory.generate_report()
```

```python
class Contact:
    """è¯çµ¡äººé¡åˆ¥"""
    
    def __init__(self, name, phone, email, address=None):
        """åˆå§‹åŒ–è¯çµ¡äºº"""
        self.name = name
        self.phone = phone
        self.email = email
        self.address = address
    
    def update_phone(self, new_phone):
        """æ›´æ–°é›»è©±"""
        self.phone = new_phone
        print(f"{self.name} çš„é›»è©±å·²æ›´æ–°ç‚º {new_phone}")
    
    def update_email(self, new_email):
        """æ›´æ–°ä¿¡ç®±"""
        self.email = new_email
        print(f"{self.name} çš„ä¿¡ç®±å·²æ›´æ–°ç‚º {new_email}")
    
    def display(self):
        """é¡¯ç¤ºè¯çµ¡äººè³‡è¨Š"""
        print(f"å§“åï¼š{self.name}")
        print(f"é›»è©±ï¼š{self.phone}")
        print(f"ä¿¡ç®±ï¼š{self.email}")
        if self.address:
            print(f"åœ°å€ï¼š{self.address}")


class ContactManager:
    """è¯çµ¡äººç®¡ç†å™¨"""
    
    def __init__(self):
        """åˆå§‹åŒ–ç®¡ç†å™¨"""
        self.contacts = []
    
    def add_contact(self, contact):
        """æ–°å¢è¯çµ¡äºº"""
        self.contacts.append(contact)
        print(f"å·²æ–°å¢è¯çµ¡äººï¼š{contact.name}")
    
    def remove_contact(self, name):
        """åˆªé™¤è¯çµ¡äºº"""
        for contact in self.contacts:
            if contact.name == name:
                self.contacts.remove(contact)
                print(f"å·²åˆªé™¤è¯çµ¡äººï¼š{name}")
                return
        print(f"æ‰¾ä¸åˆ°è¯çµ¡äººï¼š{name}")
    
    def search_by_name(self, name):
        """ä¾å§“åæœå°‹ï¼ˆéƒ¨åˆ†æ¯”å°ï¼‰"""
        results = [c for c in self.contacts if name in c.name]
        if results:
            print(f"æ‰¾åˆ° {len(results)} ä½è¯çµ¡äººï¼š")
            for contact in results:
                print(f"  - {contact.name}")
        else:
            print(f"æ‰¾ä¸åˆ°å§“ååŒ…å« '{name}' çš„è¯çµ¡äºº")
        return results
    
    def search_by_phone(self, phone):
        """ä¾é›»è©±æœå°‹"""
        for contact in self.contacts:
            if contact.phone == phone:
                print(f"æ‰¾åˆ°è¯çµ¡äººï¼š{contact.name}")
                return contact
        print(f"æ‰¾ä¸åˆ°é›»è©±ï¼š{phone}")
        return None
    
    def list_all(self):
        """åˆ—å‡ºæ‰€æœ‰è¯çµ¡äºº"""
        print(f"\næ‰€æœ‰è¯çµ¡äººï¼ˆå…± {len(self.contacts)} ä½ï¼‰ï¼š")
        print("-" * 50)
        for i, contact in enumerate(self.contacts, 1):
            print(f"{i}. {contact.name} - {contact.phone}")
        print("-" * 50)
    
    def get_count(self):
        """å–å¾—è¯çµ¡äººç¸½æ•¸"""
        return len(self.contacts)


# æ¸¬è©¦
manager = ContactManager()
manager.add_contact(Contact("å¼µä¸‰", "0912-345-678", "zhang@email.com"))
manager.add_contact(Contact("æå››", "0923-456-789", "li@email.com"))
manager.add_contact(Contact("ç‹äº”", "0934-567-890", "wang@email.com", "å°åŒ—å¸‚"))

print()
result = manager.search_by_name("å¼µ")
if result:
    print()
    result[0].display()

manager.list_all()
print(f"\nç¸½è¯çµ¡äººæ•¸ï¼š{manager.get_count()}")
```

### ç·´ç¿’ 18ï¼šTodoList å¾…è¾¦äº‹é …æ‡‰ç”¨ - è§£ç­”

```python
class Task:
    """ä»»å‹™é¡åˆ¥"""
    
    task_id_counter = 0  # é¡åˆ¥å±¬æ€§ï¼šID è¨ˆæ•¸å™¨
    
    def __init__(self, title, priority):
        """åˆå§‹åŒ–ä»»å‹™"""
        Task.task_id_counter += 1
        self.task_id = Task.task_id_counter
        self.title = title
        self.priority = priority
        self.is_completed = False
    
    def complete(self):
        """æ¨™è¨˜ç‚ºå®Œæˆ"""
        self.is_completed = True
    
    def display(self):
        """é¡¯ç¤ºä»»å‹™è³‡è¨Š"""
        status = "âœ“" if self.is_completed else " "
        print(f"[{status}] {self.task_id}. {self.title} ({self.priority})")


class TodoList:
    """å¾…è¾¦äº‹é …åˆ—è¡¨"""
    
    def __init__(self):
        """åˆå§‹åŒ–å¾…è¾¦åˆ—è¡¨"""
        self.tasks = []
    
    def add_task(self, title, priority):
        """æ–°å¢ä»»å‹™"""
        task = Task(title, priority)
        self.tasks.append(task)
        print(f"å·²æ–°å¢ä»»å‹™ #{task.task_id}ï¼š{title}")
    
    def complete_task(self, task_id):
        """å®Œæˆä»»å‹™"""
        for task in self.tasks:
            if task.task_id == task_id:
                task.complete()
                print(f"ä»»å‹™ #{task_id} å·²å®Œæˆ")
                return
        print(f"æ‰¾ä¸åˆ°ä»»å‹™ #{task_id}")
    
    def remove_task(self, task_id):
        """åˆªé™¤ä»»å‹™"""
        for task in self.tasks:
            if task.task_id == task_id:
                self.tasks.remove(task)
                print(f"ä»»å‹™ #{task_id} å·²åˆªé™¤")
                return
        print(f"æ‰¾ä¸åˆ°ä»»å‹™ #{task_id}")
    
    def list_pending(self):
        """åˆ—å‡ºæœªå®Œæˆä»»å‹™"""
        pending = [t for t in self.tasks if not t.is_completed]
        print(f"\nå¾…å®Œæˆä»»å‹™ï¼ˆå…± {len(pending)} é …ï¼‰ï¼š")
        for task in pending:
            task.display()
    
    def list_completed(self):
        """åˆ—å‡ºå·²å®Œæˆä»»å‹™"""
        completed = [t for t in self.tasks if t.is_completed]
        print(f"\nå·²å®Œæˆä»»å‹™ï¼ˆå…± {len(completed)} é …ï¼‰ï¼š")
        for task in completed:
            task.display()
    
    def filter_by_priority(self, priority):
        """ä¾å„ªå…ˆç´šéæ¿¾"""
        filtered = [t for t in self.tasks if t.priority == priority]
        print(f"\nå„ªå…ˆç´š '{priority}' çš„ä»»å‹™ï¼ˆå…± {len(filtered)} é …ï¼‰ï¼š")
        for task in filtered:
            task.display()
        return filtered
    
    def get_statistics(self):
        """é¡¯ç¤ºçµ±è¨ˆ"""
        total = len(self.tasks)
        completed = sum(1 for t in self.tasks if t.is_completed)
        pending = total - completed
        print(f"\nç¸½ä»»å‹™æ•¸ï¼š{total}")
        print(f"å·²å®Œæˆï¼š{completed}")
        print(f"å¾…å®Œæˆï¼š{pending}")


# æ¸¬è©¦
todo = TodoList()
todo.add_task("å®Œæˆ Python ä½œæ¥­", "é«˜")
todo.add_task("è²·æ™šé¤", "ä¸­")
todo.add_task("å›è¦†ä¿¡ä»¶", "ä½")

print()
todo.complete_task(1)

todo.list_pending()
todo.list_completed()

todo.filter_by_priority("é«˜")

todo.get_statistics()
```
