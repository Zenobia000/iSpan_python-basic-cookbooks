```python
class Employee:
    """å“¡å·¥åŸºé¡"""
    
    def __init__(self, name, emp_id, salary):
        self.name = name
        self.emp_id = emp_id
        self.salary = salary
    
    def get_total_salary(self):
        """è¨ˆç®—ç¸½è–ªè³‡"""
        return self.salary
    
    def display_info(self):
        """é¡¯ç¤ºå“¡å·¥è³‡è¨Š"""
        print(f"å§“å: {self.name}")
        print(f"å“¡å·¥ç·¨è™Ÿ: {self.emp_id}")
        print(f"è–ªè³‡: ${self.get_total_salary():,.2f}")

class Manager(Employee):
    """ç¶“ç†é¡åˆ¥ï¼ˆç¹¼æ‰¿ Employeeï¼‰"""
    
    def __init__(self, name, emp_id, salary, bonus):
        # å‘¼å«çˆ¶é¡åˆ¥çš„å»ºæ§‹å­
        super().__init__(name, emp_id, salary)
        self.bonus = bonus
    
    def get_total_salary(self):
        """è¦†å¯«æ–¹æ³•ï¼šç¸½è–ªè³‡ = åŸºæœ¬è–ªè³‡ + çé‡‘"""
        return self.salary + self.bonus
    
    def display_info(self):
        """æ“´å±•çˆ¶é¡åˆ¥æ–¹æ³•"""
        super().display_info()  # å…ˆå‘¼å«çˆ¶é¡åˆ¥æ–¹æ³•
        print(f"çé‡‘: ${self.bonus:,.2f}")

# æ¸¬è©¦
print("=== ä¸€èˆ¬å“¡å·¥ ===")
emp = Employee("Alice", "E001", 50000)
emp.display_info()

print("\n=== ç¶“ç† ===")
mgr = Manager("Bob", "M001", 80000, 20000)
mgr.display_info()

# é©—è­‰ç¹¼æ‰¿é—œä¿‚
print(f"\nmgr æ˜¯ Manager çš„å¯¦ä¾‹å—ï¼Ÿ{isinstance(mgr, Manager)}")
print(f"mgr æ˜¯ Employee çš„å¯¦ä¾‹å—ï¼Ÿ{isinstance(mgr, Employee)}")
```

### è§£é¡Œæ€è·¯
- å®šç¾©å…±åŒä»‹é¢ï¼ˆarea, perimeterï¼‰
- æ¯å€‹å­é¡åˆ¥å¯¦ä½œè‡ªå·±çš„è¨ˆç®—é‚è¼¯
- ç”¨å¤šå‹è™•ç†ç•°è³ªé›†åˆ

```python
import math

class Shape:
    """å½¢ç‹€åŸºé¡"""
    
    def area(self):
        """è¨ˆç®—é¢ç©ï¼ˆå­é¡åˆ¥éœ€è¦†å¯«ï¼‰"""
        raise NotImplementedError("å­é¡åˆ¥å¿…é ˆå¯¦ä½œ area() æ–¹æ³•")
    
    def perimeter(self):
        """è¨ˆç®—å‘¨é•·ï¼ˆå­é¡åˆ¥éœ€è¦†å¯«ï¼‰"""
        raise NotImplementedError("å­é¡åˆ¥å¿…é ˆå¯¦ä½œ perimeter() æ–¹æ³•")
    
    def display(self):
        """é¡¯ç¤ºè³‡è¨Š"""
        print(f"{self.__class__.__name__}:")
        print(f"  é¢ç©: {self.area():.2f}")
        print(f"  å‘¨é•·: {self.perimeter():.2f}")

class Circle(Shape):
    """åœ“å½¢"""
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    """çŸ©å½¢"""
    
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

class Triangle(Shape):
    """ä¸‰è§’å½¢"""
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def area(self):
        # ä½¿ç”¨æµ·é¾å…¬å¼
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
    
    def perimeter(self):
        return self.a + self.b + self.c

# å¤šå‹çš„å¨åŠ›ï¼šç”¨åŒä¸€å€‹å‡½å¼è™•ç†ä¸åŒé¡å‹
def calculate_total_area(shapes):
    """è¨ˆç®—æ‰€æœ‰åœ–å½¢çš„ç¸½é¢ç©"""
    total = sum(shape.area() for shape in shapes)
    return total

# æ¸¬è©¦
shapes = [
    Circle(5),
    Rectangle(4, 6),
    Triangle(3, 4, 5)
]

print("=== å„åœ–å½¢è³‡è¨Š ===")
for shape in shapes:
    shape.display()
    print()

print(f"ç¸½é¢ç©: {calculate_total_area(shapes):.2f}")
```

## ç¯„ä¾‹ 3ï¼šå¤šé‡ç¹¼æ‰¿èˆ‡ MRO

### å•é¡Œæè¿°
è¨­è¨ˆä¸€å€‹æ™ºæ…§å‹è£ç½®ç³»çµ±ï¼Œè£ç½®å¯ä»¥åŒæ™‚å…·å‚™å¤šç¨®åŠŸèƒ½ã€‚

### éœ€æ±‚
1. Cameraã€GPSã€Phone ä¸‰å€‹åŠŸèƒ½é¡åˆ¥

### è§£é¡Œæ€è·¯
- ä½¿ç”¨å¤šé‡ç¹¼æ‰¿çµ„åˆåŠŸèƒ½
- æ‰€æœ‰é¡åˆ¥éƒ½ä½¿ç”¨ super() ç¢ºä¿å”ä½œå¼ç¹¼æ‰¿
- æª¢æŸ¥ MRO ç†è§£æ–¹æ³•æŸ¥æ‰¾é †åº

```python
class Device:
    """è£ç½®åŸºé¡"""
    
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        print(f"Device.__init__: {brand} {model}")
    
    def power_on(self):
        print(f"{self.brand} {self.model} is powering on...")

class Camera(Device):
    """ç›¸æ©ŸåŠŸèƒ½"""
    
    def __init__(self, brand, model, megapixels):
        super().__init__(brand, model)
        self.megapixels = megapixels
        print(f"Camera.__init__: {megapixels}MP")
    
    def take_photo(self):
        print(f"Taking photo with {self.megapixels}MP camera")

class GPS(Device):
    """GPS åŠŸèƒ½"""
    
    def __init__(self, brand, model, accuracy):
        super().__init__(brand, model)
        self.accuracy = accuracy
        print(f"GPS.__init__: {accuracy}m accuracy")
    
    def get_location(self):
        print(f"Getting location with {self.accuracy}m accuracy")

class Phone(Device):
    """é›»è©±åŠŸèƒ½"""
    
    def __init__(self, brand, model, sim_count):
        super().__init__(brand, model)
        self.sim_count = sim_count
        print(f"Phone.__init__: {sim_count} SIM slots")
    
    def make_call(self, number):
        print(f"Calling {number}...")

class Smartphone(Camera, GPS, Phone):
    """æ™ºæ…§å‹æ‰‹æ©Ÿï¼ˆå¤šé‡ç¹¼æ‰¿ï¼‰"""
    
    def __init__(self, brand, model, megapixels, accuracy, sim_count, os):
        # æ³¨æ„ï¼šåªå‘¼å«ä¸€æ¬¡ super().__init__()
        # MRO æœƒç¢ºä¿æ‰€æœ‰çˆ¶é¡åˆ¥éƒ½è¢«åˆå§‹åŒ–
        super().__init__(brand, model, megapixels)
        # æ‰‹å‹•åˆå§‹åŒ–å…¶ä»–åŠŸèƒ½çš„ç‰¹æœ‰å±¬æ€§
        self.accuracy = accuracy
        self.sim_count = sim_count
        self.os = os
        print(f"Smartphone.__init__: OS {os}")
    
    def show_features(self):
        print(f"\n{self.brand} {self.model} Features:")
        print(f"  - {self.megapixels}MP Camera")
        print(f"  - GPS ({self.accuracy}m accuracy)")
        print(f"  - {self.sim_count} SIM slots")
        print(f"  - OS: {self.os}")

# æŸ¥çœ‹ MRO
print("=== Smartphone çš„ MRO ===")
for i, cls in enumerate(Smartphone.mro(), 1):
    print(f"{i}. {cls.__name__}")

print("\n=== å‰µå»º Smartphone å¯¦ä¾‹ ===")
phone = Smartphone("Apple", "iPhone 15", 48, 5, 2, "iOS 17")

print("\n=== æ¸¬è©¦åŠŸèƒ½ ===")
phone.show_features()
phone.take_photo()
phone.get_location()
phone.make_call("0912-345-678")
```

## ç¯„ä¾‹ 4ï¼šæŠ½è±¡åŸºé¡ï¼ˆéŠ€è¡Œå¸³æˆ¶ç³»çµ±ï¼‰

### å•é¡Œæè¿°
å»ºç«‹ä¸€å€‹éŠ€è¡Œå¸³æˆ¶ç³»çµ±ï¼Œä½¿ç”¨æŠ½è±¡åŸºé¡å®šç¾©ä»‹é¢è¦ç¯„ã€‚

### éœ€æ±‚
1. Account æŠ½è±¡åŸºé¡ï¼šå®šç¾©å­˜æ¬¾ã€ææ¬¾ä»‹é¢
2. SavingsAccountã€CheckingAccount å­é¡åˆ¥
3. ä¸åŒå¸³æˆ¶æœ‰ä¸åŒçš„åˆ©ç‡å’Œæ‰‹çºŒè²»è¦å‰‡

### è§£é¡Œæ€è·¯
- ä½¿ç”¨ ABC å®šç¾©æŠ½è±¡æ–¹æ³•
- å¼·åˆ¶å­é¡åˆ¥å¯¦ä½œç‰¹å®šæ–¹æ³•
- ç¢ºä¿ä»‹é¢ä¸€è‡´æ€§

```python
from abc import ABC, abstractmethod

class Account(ABC):
    """éŠ€è¡Œå¸³æˆ¶æŠ½è±¡åŸºé¡"""
    
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance
    
    @abstractmethod
    def deposit(self, amount):
        """å­˜æ¬¾ï¼ˆæŠ½è±¡æ–¹æ³•ï¼‰"""
        pass
    
    @abstractmethod
    def withdraw(self, amount):
        """ææ¬¾ï¼ˆæŠ½è±¡æ–¹æ³•ï¼‰"""
        pass
    
    @abstractmethod
    def calculate_interest(self):
        """è¨ˆç®—åˆ©æ¯ï¼ˆæŠ½è±¡æ–¹æ³•ï¼‰"""
        pass
    
    def get_balance(self):
        """æŸ¥è©¢é¤˜é¡ï¼ˆå…·é«”æ–¹æ³•ï¼‰"""
        return self.balance
    
    def display_info(self):
        """é¡¯ç¤ºå¸³æˆ¶è³‡è¨Š"""
        print(f"å¸³æˆ¶é¡å‹: {self.__class__.__name__}")
        print(f"å¸³è™Ÿ: {self.account_number}")
        print(f"é¤˜é¡: ${self.balance:,.2f}")

class SavingsAccount(Account):
    """å„²è“„å¸³æˆ¶"""
    
    INTEREST_RATE = 0.02  # 2% å¹´åˆ©ç‡
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"å­˜æ¬¾ ${amount:,.2f} æˆåŠŸ")
        else:
            print("å­˜æ¬¾é‡‘é¡å¿…é ˆå¤§æ–¼ 0")
    
    def withdraw(self, amount):
        if amount > 0:
            if self.balance >= amount:
                self.balance -= amount
                print(f"ææ¬¾ ${amount:,.2f} æˆåŠŸ")
            else:
                print("é¤˜é¡ä¸è¶³")
        else:
            print("ææ¬¾é‡‘é¡å¿…é ˆå¤§æ–¼ 0")
    
    def calculate_interest(self):
        interest = self.balance * self.INTEREST_RATE
        print(f"å¹´åˆ©æ¯: ${interest:,.2f}")
        return interest

class CheckingAccount(Account):
    """æ”¯ç¥¨å¸³æˆ¶"""
    
    TRANSACTION_FEE = 10  # æ¯ç­†æ‰‹çºŒè²»
    FREE_TRANSACTIONS = 5  # å…è²»äº¤æ˜“æ¬¡æ•¸
    
    def __init__(self, account_number, balance=0):
        super().__init__(account_number, balance)
        self.transaction_count = 0
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transaction_count += 1
            self._apply_fee()
            print(f"å­˜æ¬¾ ${amount:,.2f} æˆåŠŸ")
        else:
            print("å­˜æ¬¾é‡‘é¡å¿…é ˆå¤§æ–¼ 0")
    
    def withdraw(self, amount):
        if amount > 0:
            total_amount = amount
            if self.balance >= total_amount:
                self.balance -= total_amount
                self.transaction_count += 1
                self._apply_fee()
                print(f"ææ¬¾ ${amount:,.2f} æˆåŠŸ")
            else:
                print("é¤˜é¡ä¸è¶³")
        else:
            print("ææ¬¾é‡‘é¡å¿…é ˆå¤§æ–¼ 0")
    
    def calculate_interest(self):
        # æ”¯ç¥¨å¸³æˆ¶ç„¡åˆ©æ¯
        print("æ”¯ç¥¨å¸³æˆ¶ç„¡åˆ©æ¯")
        return 0
    
    def _apply_fee(self):
        """å¥—ç”¨æ‰‹çºŒè²»"""
        if self.transaction_count > self.FREE_TRANSACTIONS:
            self.balance -= self.TRANSACTION_FEE
            print(f"  (æ‰‹çºŒè²» ${self.TRANSACTION_FEE})")

# æ¸¬è©¦
print("=== å„²è“„å¸³æˆ¶ ===")
savings = SavingsAccount("SAV-001", 10000)
savings.display_info()
savings.deposit(5000)
savings.withdraw(2000)
savings.calculate_interest()
print(f"é¤˜é¡: ${savings.get_balance():,.2f}")

print("\n=== æ”¯ç¥¨å¸³æˆ¶ ===")
checking = CheckingAccount("CHK-001", 20000)
checking.display_info()
for i in range(7):  # æ¸¬è©¦æ‰‹çºŒè²»æ©Ÿåˆ¶
    checking.deposit(1000)
print(f"æœ€çµ‚é¤˜é¡: ${checking.get_balance():,.2f}")
```
