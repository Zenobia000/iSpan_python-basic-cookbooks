````python
# å®šç¾©çˆ¶é¡ Animal
class Animal:
    """å‹•ç‰©é¡åˆ¥ï¼ˆçˆ¶é¡ï¼‰"""
    
    def __init__(self, name, age):
        """å»ºæ§‹å­"""
        self.name = name
        self.age = age
    
    def eat(self):
        """é€²é£Ÿæ–¹æ³•"""
        return f"{self.name} is eating"
    
    def sleep(self):
        """ç¡è¦ºæ–¹æ³•"""
        return f"{self.name} is sleeping"
    
    def info(self):
        """é¡¯ç¤ºè³‡è¨Š"""
        return f"{self.name} is {self.age} years old"


# å®šç¾©å­é¡ Dogï¼ˆç¹¼æ‰¿ Animalï¼‰
class Dog(Animal):
    """ç‹—é¡åˆ¥ï¼ˆå­é¡ï¼‰"""
    
    def bark(self):
        """ç‹—ç‰¹æœ‰çš„å å«æ–¹æ³•"""
        return f"{self.name} says Woof!"


# å®šç¾©å­é¡ Catï¼ˆç¹¼æ‰¿ Animalï¼‰
class Cat(Animal):
    """è²“é¡åˆ¥ï¼ˆå­é¡ï¼‰"""
    
    def meow(self):
        """è²“ç‰¹æœ‰çš„å«è²æ–¹æ³•"""
        return f"{self.name} says Meow!"


# æ¸¬è©¦ç¨‹å¼ç¢¼
print("=== å»ºç«‹ç‰©ä»¶ ===")
dog = Dog("Buddy", 3)
cat = Cat("Whiskers", 2)

print("\n=== ä½¿ç”¨ç¹¼æ‰¿ä¾†çš„æ–¹æ³• ===")
print(dog.info())    # Buddy is 3 years oldï¼ˆç¹¼æ‰¿è‡ª Animalï¼‰
print(dog.eat())     # Buddy is eatingï¼ˆç¹¼æ‰¿è‡ª Animalï¼‰
print(dog.sleep())   # Buddy is sleepingï¼ˆç¹¼æ‰¿è‡ª Animalï¼‰

print("\n=== ä½¿ç”¨å­é¡ç‰¹æœ‰çš„æ–¹æ³• ===")
print(dog.bark())    # Buddy says Woof!ï¼ˆDog ç‰¹æœ‰ï¼‰
print(cat.meow())    # Whiskers says Meow!ï¼ˆCat ç‰¹æœ‰ï¼‰

print("\n=== æª¢æŸ¥ç¹¼æ‰¿é—œä¿‚ ===")
print(f"dog æ˜¯ Dog çš„å¯¦ä¾‹å—ï¼Ÿ {isinstance(dog, Dog)}")        # True
print(f"dog æ˜¯ Animal çš„å¯¦ä¾‹å—ï¼Ÿ {isinstance(dog, Animal)}")  # True
print(f"dog æ˜¯ Cat çš„å¯¦ä¾‹å—ï¼Ÿ {isinstance(dog, Cat)}")        # False

print(f"\nDog æ˜¯ Animal çš„å­é¡å—ï¼Ÿ {issubclass(Dog, Animal)}")  # True
print(f"Cat æ˜¯ Animal çš„å­é¡å—ï¼Ÿ {issubclass(Cat, Animal)}")    # True
print(f"Dog æ˜¯ Cat çš„å­é¡å—ï¼Ÿ {issubclass(Dog, Cat)}")          # False
````

### ç¯„ä¾‹ 2ï¼šsuper() çš„ä½¿ç”¨ï¼ˆå‘¼å«çˆ¶é¡æ–¹æ³•ï¼‰

````python
# çˆ¶é¡ï¼šEmployeeï¼ˆå“¡å·¥ï¼‰
class Employee:
    """å“¡å·¥åŸºé¡"""
    
    def __init__(self, name, employee_id, salary):
        """å»ºæ§‹å­"""
        self.name = name
        self.employee_id = employee_id
        self.salary = salary
    
    def display_info(self):
        """é¡¯ç¤ºå“¡å·¥è³‡è¨Š"""
        return f"å“¡å·¥ï¼š{self.name}ï¼ŒIDï¼š{self.employee_id}ï¼Œè–ªè³‡ï¼š${self.salary}"
    
    def give_raise(self, amount):
        """åŠ è–ª"""
        self.salary += amount
        return f"{self.name} åŠ è–ª ${amount}ï¼Œæ–°è–ªè³‡ï¼š${self.salary}"


# å­é¡ï¼šManagerï¼ˆç¶“ç†ï¼‰
class Manager(Employee):
    """ç¶“ç†é¡åˆ¥ï¼ˆç¹¼æ‰¿ Employeeï¼‰"""
    
    def __init__(self, name, employee_id, salary, department):
        """å»ºæ§‹å­ - ä½¿ç”¨ super() å‘¼å«çˆ¶é¡å»ºæ§‹å­"""
        # å‘¼å«çˆ¶é¡çš„ __init__ï¼Œåˆå§‹åŒ– name, employee_id, salary
        super().__init__(name, employee_id, salary)
        # æ·»åŠ  Manager ç‰¹æœ‰çš„å±¬æ€§
        self.department = department
    
    def display_info(self):
        """è¦†å¯«çˆ¶é¡æ–¹æ³• - æ“´å±•åŠŸèƒ½"""
        # å‘¼å«çˆ¶é¡çš„ display_info()
        parent_info = super().display_info()
        # æ·»åŠ éƒ¨é–€è³‡è¨Š
        return f"{parent_info}ï¼Œéƒ¨é–€ï¼š{self.department}"
    
    def manage_team(self):
        """Manager ç‰¹æœ‰çš„æ–¹æ³•"""
        return f"{self.name} æ­£åœ¨ç®¡ç† {self.department} éƒ¨é–€"


# æ¸¬è©¦ç¨‹å¼ç¢¼
print("=== å»ºç«‹ç‰©ä»¶ ===")
emp = Employee("Alice", "E001", 50000)
mgr = Manager("Bob", "M001", 80000, "Engineering")

print("\n=== Employee ç‰©ä»¶ ===")
print(emp.display_info())
print(emp.give_raise(5000))

print("\n=== Manager ç‰©ä»¶ ===")
print(mgr.display_info())  # ä½¿ç”¨æ“´å±•å¾Œçš„æ–¹æ³•
print(mgr.give_raise(10000))  # ç¹¼æ‰¿è‡ª Employee
print(mgr.manage_team())      # Manager ç‰¹æœ‰æ–¹æ³•

print("\n=== æŸ¥çœ‹å±¬æ€§ ===")
print(f"Manager çš„å±¬æ€§ï¼š{vars(mgr)}")
````

### ç¯„ä¾‹ 3ï¼šæ–¹æ³•è¦†å¯«èˆ‡å¤šå‹ï¼ˆåœ–å½¢é¢ç©è¨ˆç®—ï¼‰

````python
import math

# çˆ¶é¡ï¼šShapeï¼ˆåœ–å½¢ï¼‰
class Shape:
    """åœ–å½¢åŸºé¡"""
    
    def __init__(self, name):
        self.name = name
    
    def area(self):
        """è¨ˆç®—é¢ç© - çˆ¶é¡æä¾›é è¨­å¯¦ä½œ"""
        return 0
    
    def perimeter(self):
        """è¨ˆç®—å‘¨é•· - çˆ¶é¡æä¾›é è¨­å¯¦ä½œ"""
        return 0
    
    def describe(self):
        """æè¿°åœ–å½¢"""
        return f"{self.name} - é¢ç©ï¼š{self.area():.2f}ï¼Œå‘¨é•·ï¼š{self.perimeter():.2f}"


# å­é¡ï¼šRectangleï¼ˆçŸ©å½¢ï¼‰
class Rectangle(Shape):
    """çŸ©å½¢é¡åˆ¥"""
    
    def __init__(self, width, height):
        super().__init__("çŸ©å½¢")
        self.width = width
        self.height = height
    
    def area(self):
        """è¦†å¯« area() - çŸ©å½¢é¢ç© = é•· Ã— å¯¬"""
        return self.width * self.height
    
    def perimeter(self):
        """è¦†å¯« perimeter() - çŸ©å½¢å‘¨é•· = 2(é•· + å¯¬)"""
        return 2 * (self.width + self.height)


# å­é¡ï¼šCircleï¼ˆåœ“å½¢ï¼‰
class Circle(Shape):
    """åœ“å½¢é¡åˆ¥"""
    
    def __init__(self, radius):
        super().__init__("åœ“å½¢")
        self.radius = radius
    
    def area(self):
        """è¦†å¯« area() - åœ“å½¢é¢ç© = Ï€ Ã— rÂ²"""
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        """è¦†å¯« perimeter() - åœ“å½¢å‘¨é•· = 2 Ã— Ï€ Ã— r"""
        return 2 * math.pi * self.radius


# å­é¡ï¼šTriangleï¼ˆä¸‰è§’å½¢ï¼‰
class Triangle(Shape):
    """ä¸‰è§’å½¢é¡åˆ¥"""
    
    def __init__(self, a, b, c):
        super().__init__("ä¸‰è§’å½¢")
        self.a = a  # é‚Š a
        self.b = b  # é‚Š b
        self.c = c  # é‚Š c
    
    def area(self):
        """è¦†å¯« area() - ä½¿ç”¨æµ·é¾å…¬å¼è¨ˆç®—é¢ç©"""
        s = (self.a + self.b + self.c) / 2  # åŠå‘¨é•·
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
    
    def perimeter(self):
        """è¦†å¯« perimeter() - ä¸‰è§’å½¢å‘¨é•· = a + b + c"""
        return self.a + self.b + self.c


# æ¸¬è©¦å¤šå‹
print("=== å»ºç«‹åœ–å½¢ç‰©ä»¶ ===")
shapes = [
    Rectangle(5, 3),
    Circle(4),
    Triangle(3, 4, 5)
]

print("\n=== å¤šå‹ç¤ºç¯„ï¼šçµ±ä¸€ä»‹é¢è™•ç†ä¸åŒåœ–å½¢ ===")
for shape in shapes:
    print(shape.describe())  # åŒæ¨£å‘¼å« describe()ï¼Œä½†å…§éƒ¨ area() è¡Œç‚ºä¸åŒ

print("\n=== è¨ˆç®—ç¸½é¢ç© ===")
total_area = sum(shape.area() for shape in shapes)
print(f"æ‰€æœ‰åœ–å½¢çš„ç¸½é¢ç©ï¼š{total_area:.2f}")

print("\n=== æ‰¾å‡ºé¢ç©æœ€å¤§çš„åœ–å½¢ ===")
max_shape = max(shapes, key=lambda s: s.area())
print(f"é¢ç©æœ€å¤§ï¼š{max_shape.name}ï¼Œé¢ç© = {max_shape.area():.2f}")
````

````python
# é‘½çŸ³ç¹¼æ‰¿ç¯„ä¾‹
class A:
    """é ‚å±¤é¡åˆ¥ A"""
    
    def method(self):
        return "A.method()"
    
    def who_am_i(self):
        return "I am A"


class B(A):
    """B ç¹¼æ‰¿ A"""
    
    def method(self):
        return "B.method()"
    
    def who_am_i(self):
        return "I am B"


class C(A):
    """C ç¹¼æ‰¿ A"""
    
    def method(self):
        return "C.method()"
    
    def who_am_i(self):
        return "I am C"


class D(B, C):
    """D ç¹¼æ‰¿ B å’Œ C - é‘½çŸ³ç¹¼æ‰¿
    
    ç¹¼æ‰¿åœ–ï¼š
        A
       / \
      B   C
       \ /
        D
    """
    pass


# æ¸¬è©¦ MRO
print("=== é‘½çŸ³ç¹¼æ‰¿ ===")
d = D()

print("\n=== å‘¼å« method() - æœƒç”¨å“ªå€‹ï¼Ÿ ===")
print(d.method())      # "B.method()" - ä¾æ“š MROï¼ŒB å„ªå…ˆ
print(d.who_am_i())    # "I am B" - ä¾æ“š MROï¼ŒB å„ªå…ˆ

print("\n=== æŸ¥çœ‹ MROï¼ˆMethod Resolution Orderï¼‰===")
print("ä½¿ç”¨ __mro__:")
print(D.__mro__)
# (<class 'D'>, <class 'B'>, <class 'C'>, <class 'A'>, <class 'object'>)

print("\nä½¿ç”¨ .mro():")
for cls in D.mro():
    print(f"  -> {cls.__name__}")
# D -> B -> C -> A -> object

print("\n=== MRO è¦å‰‡èªªæ˜ ===")
print("""
MRO éµå¾ª C3 ç·šæ€§åŒ–æ¼”ç®—æ³•ï¼š
1. å­é¡å„ªå…ˆæ–¼çˆ¶é¡ï¼ˆD > B, C > Aï¼‰
2. å¤šå€‹çˆ¶é¡æŒ‰ç¹¼æ‰¿é †åºï¼ˆB > Cï¼Œå› ç‚º class D(B, C)ï¼‰
3. æ¯å€‹é¡åˆ¥åªå‡ºç¾ä¸€æ¬¡
4. çµæœï¼šD -> B -> C -> A -> object
""")


# super() èˆ‡ MRO çš„é—œä¿‚
class E(A):
    def method(self):
        result = super().method()  # ä¾æ“š MRO æ‰¾ä¸‹ä¸€å€‹é¡åˆ¥
        return f"E.method() -> {result}"

class F(E, C):
    def method(self):
        result = super().method()  # ä¾æ“š MRO æ‰¾ä¸‹ä¸€å€‹é¡åˆ¥
        return f"F.method() -> {result}"

print("\n=== super() èˆ‡ MRO ===")
f = F()
print(f.method())
# F.method() -> E.method() -> C.method()

print("\nF çš„ MRO:")
for cls in F.mro():
    print(f"  -> {cls.__name__}")
# F -> E -> C -> A -> object


# å¯¦ç”¨æŠ€å·§ï¼šæª¢æŸ¥æ–¹æ³•ä¾†æº
print("\n=== å¯¦ç”¨æŠ€å·§ï¼šæª¢æŸ¥æ–¹æ³•ä¾†æº ===")
import inspect

def find_method_source(obj, method_name):
    """æ‰¾å‡ºæ–¹æ³•å®šç¾©åœ¨å“ªå€‹é¡åˆ¥"""
    method = getattr(obj, method_name)
    return inspect.getmro(type(obj))[0].__name__, method.__qualname__

print(f"d.method() ä¾†è‡ªï¼š{find_method_source(d, 'method')}")
print(f"f.method() ä¾†è‡ªï¼š{find_method_source(f, 'method')}")
````

````python
from abc import ABC, abstractmethod
from datetime import datetime

# æŠ½è±¡åŸºé¡ï¼šPaymentMethodï¼ˆæ”¯ä»˜æ–¹å¼ï¼‰
class PaymentMethod(ABC):
    """æ”¯ä»˜æ–¹å¼æŠ½è±¡åŸºé¡ - å®šç¾©ä»‹é¢"""
    
    def __init__(self, user_name):
        self.user_name = user_name
        self.transaction_history = []
    
    @abstractmethod
    def process_payment(self, amount):
        """è™•ç†æ”¯ä»˜ - å­é¡å¿…é ˆå¯¦ä½œ"""
        pass
    
    @abstractmethod
    def refund(self, transaction_id):
        """é€€æ¬¾ - å­é¡å¿…é ˆå¯¦ä½œ"""
        pass
    
    def record_transaction(self, transaction_id, amount, status):
        """è¨˜éŒ„äº¤æ˜“ - å…±ç”¨æ–¹æ³•ï¼ˆéæŠ½è±¡ï¼‰"""
        record = {
            'id': transaction_id,
            'amount': amount,
            'status': status,
            'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        }
        self.transaction_history.append(record)
        return record
    
    def get_history(self):
        """å–å¾—äº¤æ˜“è¨˜éŒ„ - å…±ç”¨æ–¹æ³•ï¼ˆéæŠ½è±¡ï¼‰"""
        return self.transaction_history


# å…·é«”é¡åˆ¥ï¼šCreditCardï¼ˆä¿¡ç”¨å¡ï¼‰
class CreditCard(PaymentMethod):
    """ä¿¡ç”¨å¡æ”¯ä»˜"""
    
    def __init__(self, user_name, card_number, cvv):
        super().__init__(user_name)
        self.card_number = card_number
        self.cvv = cvv
    
    def process_payment(self, amount):
        """å¯¦ä½œ process_payment"""
        transaction_id = f"CC-{len(self.transaction_history) + 1:04d}"
        # æ¨¡æ“¬ä¿¡ç”¨å¡æ”¯ä»˜æµç¨‹
        if amount > 0:
            self.record_transaction(transaction_id, amount, "æˆåŠŸ")
            return f"ä¿¡ç”¨å¡æ”¯ä»˜æˆåŠŸï¼š${amount}ï¼ˆäº¤æ˜“ç·¨è™Ÿï¼š{transaction_id}ï¼‰"
        else:
            self.record_transaction(transaction_id, amount, "å¤±æ•—")
            return "æ”¯ä»˜é‡‘é¡å¿…é ˆå¤§æ–¼ 0"
    
    def refund(self, transaction_id):
        """å¯¦ä½œ refund"""
        for record in self.transaction_history:
            if record['id'] == transaction_id and record['status'] == "æˆåŠŸ":
                record['status'] = "å·²é€€æ¬¾"
                return f"é€€æ¬¾æˆåŠŸï¼š${record['amount']}ï¼ˆäº¤æ˜“ç·¨è™Ÿï¼š{transaction_id}ï¼‰"
        return "æ‰¾ä¸åˆ°è©²äº¤æ˜“æˆ–å·²é€€æ¬¾"


# å…·é«”é¡åˆ¥ï¼šPayPal
class PayPal(PaymentMethod):
    """PayPal æ”¯ä»˜"""
    
    def __init__(self, user_name, email):
        super().__init__(user_name)
        self.email = email
    
    def process_payment(self, amount):
        """å¯¦ä½œ process_payment"""
        transaction_id = f"PP-{len(self.transaction_history) + 1:04d}"
        # æ¨¡æ“¬ PayPal æ”¯ä»˜æµç¨‹
        if amount > 0:
            self.record_transaction(transaction_id, amount, "æˆåŠŸ")
            return f"PayPal æ”¯ä»˜æˆåŠŸï¼š${amount}ï¼ˆäº¤æ˜“ç·¨è™Ÿï¼š{transaction_id}ï¼ŒEmailï¼š{self.email}ï¼‰"
        else:
            self.record_transaction(transaction_id, amount, "å¤±æ•—")
            return "æ”¯ä»˜é‡‘é¡å¿…é ˆå¤§æ–¼ 0"
    
    def refund(self, transaction_id):
        """å¯¦ä½œ refund"""
        for record in self.transaction_history:
            if record['id'] == transaction_id and record['status'] == "æˆåŠŸ":
                record['status'] = "å·²é€€æ¬¾"
                return f"é€€æ¬¾æˆåŠŸï¼š${record['amount']}ï¼ˆäº¤æ˜“ç·¨è™Ÿï¼š{transaction_id}ï¼Œé€€è‡³ {self.email}ï¼‰"
        return "æ‰¾ä¸åˆ°è©²äº¤æ˜“æˆ–å·²é€€æ¬¾"


# å…·é«”é¡åˆ¥ï¼šCryptocurrencyï¼ˆåŠ å¯†è²¨å¹£ï¼‰
class Cryptocurrency(PaymentMethod):
    """åŠ å¯†è²¨å¹£æ”¯ä»˜"""
    
    def __init__(self, user_name, wallet_address):
        super().__init__(user_name)
        self.wallet_address = wallet_address
    
    def process_payment(self, amount):
        """å¯¦ä½œ process_payment"""
        transaction_id = f"CRYPTO-{len(self.transaction_history) + 1:04d}"
        # æ¨¡æ“¬åŠ å¯†è²¨å¹£æ”¯ä»˜æµç¨‹
        if amount > 0:
            self.record_transaction(transaction_id, amount, "æˆåŠŸ")
            return f"åŠ å¯†è²¨å¹£æ”¯ä»˜æˆåŠŸï¼š${amount}ï¼ˆäº¤æ˜“ç·¨è™Ÿï¼š{transaction_id}ï¼‰"
        else:
            self.record_transaction(transaction_id, amount, "å¤±æ•—")
            return "æ”¯ä»˜é‡‘é¡å¿…é ˆå¤§æ–¼ 0"
    
    def refund(self, transaction_id):
        """å¯¦ä½œ refund"""
        return "åŠ å¯†è²¨å¹£äº¤æ˜“ç„¡æ³•é€€æ¬¾ï¼ˆå€å¡Šéˆä¸å¯é€†ï¼‰"


# æ¸¬è©¦æŠ½è±¡åŸºé¡
print("=== æ¸¬è©¦æŠ½è±¡åŸºé¡ ===")

# å˜—è©¦å¯¦ä¾‹åŒ–æŠ½è±¡åŸºé¡ï¼ˆæœƒå ±éŒ¯ï¼‰
try:
    payment = PaymentMethod("Alice")
except TypeError as e:
    print(f"éŒ¯èª¤ï¼š{e}")
    # Can't instantiate abstract class PaymentMethod with abstract methods

print("\n=== å»ºç«‹å…·é«”æ”¯ä»˜æ–¹å¼ ===")
credit_card = CreditCard("Alice", "1234-5678-9012-3456", "123")
paypal = PayPal("Bob", "bob@example.com")
crypto = Cryptocurrency("Charlie", "0x1234567890abcdef")

# å¤šå‹ï¼šçµ±ä¸€ä»‹é¢è™•ç†ä¸åŒæ”¯ä»˜æ–¹å¼
payment_methods = [credit_card, paypal, crypto]

print("\n=== çµ±ä¸€ä»‹é¢è™•ç†æ”¯ä»˜ ===")
for i, payment in enumerate(payment_methods, 1):
    print(f"\næ”¯ä»˜æ–¹å¼ {i}: {payment.__class__.__name__}")
    print(payment.process_payment(100))
    print(payment.process_payment(50))

print("\n=== æ¸¬è©¦é€€æ¬¾ ===")
print(credit_card.refund("CC-0001"))
print(paypal.refund("PP-0001"))
print(crypto.refund("CRYPTO-0001"))  # åŠ å¯†è²¨å¹£ç„¡æ³•é€€æ¬¾

print("\n=== æŸ¥çœ‹äº¤æ˜“è¨˜éŒ„ ===")
print(f"\n{credit_card.user_name} çš„äº¤æ˜“è¨˜éŒ„ï¼š")
for record in credit_card.get_history():
    print(f"  {record}")

print("\n=== æª¢æŸ¥æŠ½è±¡æ–¹æ³• ===")
print(f"PaymentMethod çš„æŠ½è±¡æ–¹æ³•ï¼š{PaymentMethod.__abstractmethods__}")
# frozenset({'process_payment', 'refund'})
````
