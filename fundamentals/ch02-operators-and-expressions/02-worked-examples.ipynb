````python
# æ”æ°æº«åº¦è½‰è¯æ°
celsius = 25

# æ­¥é©Ÿ 1ï¼šæº«åº¦è½‰æ›
fahrenheit = celsius * 9 / 5 + 32
print(f"æ”æ°æº«åº¦: {celsius}Â°C")
print(f"è¯æ°æº«åº¦: {fahrenheit}Â°F")

# æ­¥é©Ÿ 2ï¼šåˆ¤æ–·å¤©æ°£
is_freezing = celsius <= 0
is_cold = 0 < celsius <= 10
is_cool = 10 < celsius <= 20
is_comfortable = 20 < celsius <= 28
is_hot = celsius > 28

print(f"\nå¤©æ°£ç‹€æ³ï¼š")
print(f"çµå†° (â‰¤ 0Â°C): {is_freezing}")
print(f"å¯’å†· (0-10Â°C): {is_cold}")
print(f"æ¶¼çˆ½ (10-20Â°C): {is_cool}")
print(f"èˆ’é© (20-28Â°C): {is_comfortable}")
print(f"ç‚ç†± (> 28Â°C): {is_hot}")
````

## ç¯„ä¾‹ 2ï¼šBMI è¨ˆç®—èˆ‡åˆ†é¡

### ğŸ“ å•é¡Œæè¿°
è¨ˆç®— BMIï¼ˆèº«é«”è³ªé‡æŒ‡æ•¸ï¼‰ä¸¦åˆ¤æ–·å¥åº·ç‹€æ³ã€‚

**BMI å…¬å¼**ï¼š`BMI = é«”é‡(kg) / èº«é«˜(m)Â²`

**åˆ†é¡æ¨™æº–**ï¼š
- é«”é‡éè¼•ï¼šBMI < 18.5
- æ­£å¸¸ç¯„åœï¼š18.5 â‰¤ BMI < 24
- éé‡ï¼š24 â‰¤ BMI < 27
- è‚¥èƒ–ï¼šBMI â‰¥ 27

### ğŸ¤” è§£é¡Œæ€è·¯
1. ä½¿ç”¨ `**` é‹ç®—å­è¨ˆç®—å¹³æ–¹
2. ä½¿ç”¨ `/` é‹ç®—å­è¨ˆç®— BMI
3. ä½¿ç”¨é€£é–æ¯”è¼ƒåˆ¤æ–·ç¯„åœ

````python
# BMI è¨ˆç®—
weight = 70  # å…¬æ–¤
height = 1.75  # å…¬å°º

# æ­¥é©Ÿ 1ï¼šè¨ˆç®— BMI
bmi = weight / (height ** 2)
print(f"é«”é‡: {weight} kg")
print(f"èº«é«˜: {height} m")
print(f"BMI: {bmi:.2f}")  # ä¿ç•™å…©ä½å°æ•¸

# æ­¥é©Ÿ 2ï¼šåˆ¤æ–·å¥åº·ç‹€æ³
is_underweight = bmi < 18.5
is_normal = 18.5 <= bmi < 24
is_overweight = 24 <= bmi < 27
is_obese = bmi >= 27

print(f"\nå¥åº·ç‹€æ³ï¼š")
print(f"é«”é‡éè¼• (< 18.5): {is_underweight}")
print(f"æ­£å¸¸ç¯„åœ (18.5-24): {is_normal}")
print(f"éé‡ (24-27): {is_overweight}")
print(f"è‚¥èƒ– (â‰¥ 27): {is_obese}")
````

### ğŸ” é‡é»èªªæ˜

1. **æ‹¬è™Ÿçš„é‡è¦æ€§**ï¼š
   - `weight / height ** 2` âŒ éŒ¯èª¤ï¼ˆå…ˆç®— `height ** 2`ï¼Œä½†é™¤ä»¥ weightï¼‰
   - `weight / (height ** 2)` âœ… æ­£ç¢ºï¼ˆå…ˆç®— `height ** 2`ï¼Œå†ç”¨ weight é™¤ï¼‰

2. **æŒ‡æ•¸é‹ç®—**ï¼š
   - `height ** 2` ç­‰æ–¼ `height * height`
   - `**` çš„å„ªå…ˆç´šé«˜æ–¼ `/`

3. **æ ¼å¼åŒ–è¼¸å‡º**ï¼š
   - `{bmi:.2f}` è¡¨ç¤ºæµ®é»æ•¸ä¿ç•™å…©ä½å°æ•¸

## ç¯„ä¾‹ 3ï¼šé–å¹´åˆ¤æ–·

### ğŸ“ å•é¡Œæè¿°
åˆ¤æ–·æŸå¹´æ˜¯å¦ç‚ºé–å¹´ã€‚

**é–å¹´è¦å‰‡**ï¼š
- èƒ½è¢« 4 æ•´é™¤ **ä¸”** ä¸èƒ½è¢« 100 æ•´é™¤
- **æˆ–** èƒ½è¢« 400 æ•´é™¤

**ç¯„ä¾‹**ï¼š
- 2020 æ˜¯é–å¹´ï¼ˆèƒ½è¢« 4 æ•´é™¤ï¼Œä¸èƒ½è¢« 100 æ•´é™¤ï¼‰
- 1900 ä¸æ˜¯é–å¹´ï¼ˆèƒ½è¢« 100 æ•´é™¤ï¼Œä½†ä¸èƒ½è¢« 400 æ•´é™¤ï¼‰
- 2000 æ˜¯é–å¹´ï¼ˆèƒ½è¢« 400 æ•´é™¤ï¼‰

### ğŸ¤” è§£é¡Œæ€è·¯
1. ä½¿ç”¨ `%` é‹ç®—å­æª¢æŸ¥æ•´é™¤æ€§
2. ä½¿ç”¨é‚è¼¯é‹ç®—å­çµ„åˆæ¢ä»¶
3. æ³¨æ„é‹ç®—å­å„ªå…ˆé †åºï¼š`and` å„ªå…ˆæ–¼ `or`

````python
# é–å¹´åˆ¤æ–·
year = 2024

# æ­¥é©Ÿ 1ï¼šæª¢æŸ¥å„é …æ¢ä»¶
divisible_by_4 = (year % 4 == 0)
divisible_by_100 = (year % 100 == 0)
divisible_by_400 = (year % 400 == 0)

print(f"å¹´ä»½: {year}")
print(f"èƒ½è¢« 4 æ•´é™¤: {divisible_by_4}")
print(f"èƒ½è¢« 100 æ•´é™¤: {divisible_by_100}")
print(f"èƒ½è¢« 400 æ•´é™¤: {divisible_by_400}")

# æ­¥é©Ÿ 2ï¼šåˆ¤æ–·é–å¹´
# è¦å‰‡ï¼š(èƒ½è¢«4æ•´é™¤ ä¸” ä¸èƒ½è¢«100æ•´é™¤) æˆ– èƒ½è¢«400æ•´é™¤
is_leap_year = (divisible_by_4 and not divisible_by_100) or divisible_by_400

print(f"\n{year} æ˜¯é–å¹´å—ï¼Ÿ{is_leap_year}")
````

````python
# æ¸¬è©¦å¤šå€‹å¹´ä»½
test_years = [2020, 1900, 2000, 2024, 2100]

print("é–å¹´æ¸¬è©¦ï¼š")
for year in test_years:
    is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
    print(f"{year}: {is_leap}")
````

## ç¯„ä¾‹ 4ï¼šè³¼ç‰©æŠ˜æ‰£è¨ˆç®—

### ğŸ“ å•é¡Œæè¿°
è¨ˆç®—è³¼ç‰©è»Šç¸½åƒ¹ä¸¦åˆ¤æ–·æ˜¯å¦ç¬¦åˆæŠ˜æ‰£æ¢ä»¶ã€‚

**æŠ˜æ‰£è¦å‰‡**ï¼š
- æ»¿ 1000 å…ƒæ‰“ 9 æŠ˜
- æ»¿ 3000 å…ƒæ‰“ 85 æŠ˜
- æ»¿ 5000 å…ƒæ‰“ 8 æŠ˜
- æœƒå“¡é¡å¤– 95 æŠ˜

### ğŸ¤” è§£é¡Œæ€è·¯
1. ä½¿ç”¨æ¯”è¼ƒé‹ç®—å­åˆ¤æ–·é‡‘é¡ç¯„åœ

````python
# è³¼ç‰©æŠ˜æ‰£è¨ˆç®—
original_price = 4500
is_member = True

print(f"åŸåƒ¹: ${original_price}")
print(f"æœƒå“¡: {is_member}")

# æ­¥é©Ÿ 1ï¼šåˆ¤æ–·æ»¿é¡æŠ˜æ‰£
if original_price >= 5000:
    discount_rate = 0.8
    discount_level = "8æŠ˜"
elif original_price >= 3000:
    discount_rate = 0.85
    discount_level = "85æŠ˜"
elif original_price >= 1000:
    discount_rate = 0.9
    discount_level = "9æŠ˜"
else:
    discount_rate = 1.0
    discount_level = "ç„¡æŠ˜æ‰£"

# æ­¥é©Ÿ 2ï¼šè¨ˆç®—æ»¿é¡æŠ˜æ‰£å¾Œåƒ¹æ ¼
price_after_amount_discount = original_price * discount_rate

print(f"\næ»¿é¡æŠ˜æ‰£: {discount_level}")
print(f"æŠ˜æ‰£å¾Œé‡‘é¡: ${price_after_amount_discount:.0f}")

# æ­¥é©Ÿ 3ï¼šè¨ˆç®—æœƒå“¡æŠ˜æ‰£
if is_member:
    final_price = price_after_amount_discount * 0.95
    print(f"æœƒå“¡æŠ˜æ‰£: 95æŠ˜")
else:
    final_price = price_after_amount_discount
    print(f"æœƒå“¡æŠ˜æ‰£: ç„¡")

# æ­¥é©Ÿ 4ï¼šè¨ˆç®—çœä¸‹çš„é‡‘é¡
saved = original_price - final_price

print(f"\næœ€çµ‚é‡‘é¡: ${final_price:.0f}")
print(f"çœä¸‹é‡‘é¡: ${saved:.0f} ({saved / original_price * 100:.1f}%)")
````

### ğŸ” é‡é»èªªæ˜

   - æŠ˜æ‰£ä¸èƒ½ç›´æ¥ç›¸åŠ ï¼š`0.9 + 0.95 â‰  0.85`
   - æ‡‰è©²ç›¸ä¹˜ï¼š`price * 0.9 * 0.95`

2. **æ¯”è¼ƒé‹ç®—çš„é †åº**ï¼š

## ç¯„ä¾‹ 5ï¼šæ™‚é–“è½‰æ›ï¼ˆç§’ â†’ æ™‚:åˆ†:ç§’ï¼‰

### ğŸ“ å•é¡Œæè¿°

### ğŸ¤” è§£é¡Œæ€è·¯
