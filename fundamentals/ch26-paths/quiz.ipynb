### 必須熟記的核心概念

#### 1. Path 物件建立
```python
Path('file.txt')          # 相對路徑
Path('/abs/path')         # 絕對路徑
Path.cwd()                # 當前目錄
Path.home()               # 家目錄
Path('a') / 'b' / 'c'     # 路徑組合
```

#### 2. 路徑屬性
```python
path.name                 # 完整檔名
path.stem                 # 主檔名
path.suffix               # 副檔名
path.suffixes             # 所有副檔名
path.parent               # 父目錄
path.parents              # 所有祖先
path.parts                # 路徑組成
```

#### 3. 路徑檢查
```python
path.exists()             # 是否存在
path.is_file()            # 是否為檔案
path.is_dir()             # 是否為目錄
path.is_absolute()        # 是否為絕對路徑
```

#### 4. 路徑操作
```python
path.with_suffix('.txt')  # 修改副檔名
path.with_name('new.txt') # 修改檔名
path.with_stem('name')    # 修改主檔名
path.resolve()            # 解析為絕對路徑
path.relative_to(base)    # 計算相對路徑
```

#### 5. 檔案操作
```python
path.read_text()          # 讀取文字
path.write_text('...')    # 寫入文字
path.rename(new)          # 重新命名
path.unlink()             # 刪除檔案
path.stat()               # 檔案屬性
```

#### 6. 目錄操作
```python
path.mkdir()              # 建立目錄
path.mkdir(parents=True)  # 建立父目錄
path.rmdir()              # 刪除目錄
path.iterdir()            # 遍歷目錄
```

#### 7. 模式匹配
```python
path.glob('*.txt')        # 當前層搜尋
path.rglob('*.txt')       # 遞迴搜尋
path.glob('**/*.txt')     # 等同 rglob
```

### 常見陷阱與錯誤

#### 陷阱 1：絕對路徑覆蓋
```python
# ❌ 錯誤
p = Path('base') / '/abs/path'
# 結果: /abs/path (base 被覆蓋)

# ✅ 正確
p = Path('base') / 'rel/path'
# 結果: base/rel/path
```

#### 陷阱 2：Path 不可變
```python
# ❌ 錯誤
p = Path('file.txt')
p / 'other.txt'  # p 沒有改變！

# ✅ 正確
p = Path('file.txt')
p = p / 'other.txt'  # 需要重新賦值
```

#### 陷阱 3：忘記 parents=True
```python
# ❌ 錯誤（父目錄不存在時報錯）
Path('a/b/c').mkdir()

# ✅ 正確
Path('a/b/c').mkdir(parents=True)
```

#### 陷阱 4：unlink() 錯誤處理
```python
# ❌ 不安全（檔案不存在時報錯）
path.unlink()

# ✅ 方法 1：try-except
try:
    path.unlink()
except FileNotFoundError:
    pass

# ✅ 方法 2：missing_ok (Python 3.8+)
path.unlink(missing_ok=True)
```

#### 陷阱 5：混淆 glob 與 rglob
```python
# glob() - 只搜尋當前層
list(Path('.').glob('*.txt'))  # 不含子目錄

# rglob() - 遞迴搜尋所有層
list(Path('.').rglob('*.txt'))  # 含所有子目錄
```
