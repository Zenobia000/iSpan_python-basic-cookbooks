### ç¯„ä¾‹ 7: ç›®éŒ„æ“ä½œ (mkdir, rmdir, iterdir)

**å­¸ç¿’é‡é»**: å»ºç«‹ã€åˆªé™¤ã€éæ­·ç›®éŒ„

````python
from pathlib import Path

# 1. å»ºç«‹ç›®éŒ„
print("å»ºç«‹ç›®éŒ„:")
new_dir = Path('test_directory')
new_dir.mkdir(exist_ok=True)  # exist_ok=True é¿å…å·²å­˜åœ¨æ™‚å ±éŒ¯
print(f"å»ºç«‹ç›®éŒ„: {new_dir}")

# 2. å»ºç«‹å¤šå±¤ç›®éŒ„
nested_dir = Path('level1/level2/level3')
nested_dir.mkdir(parents=True, exist_ok=True)  # parents=True å»ºç«‹çˆ¶ç›®éŒ„
print(f"å»ºç«‹å¤šå±¤ç›®éŒ„: {nested_dir}")
print()

# 3. åœ¨ç›®éŒ„ä¸­å»ºç«‹æª”æ¡ˆ
print("å»ºç«‹æ¸¬è©¦æª”æ¡ˆ:")
for i in range(3):
    file = new_dir / f'file{i}.txt'
    file.write_text(f'Content of file {i}')
    print(f"å»ºç«‹: {file.name}")

# ä¹Ÿå»ºç«‹å­ç›®éŒ„
subdir = new_dir / 'subdir'
subdir.mkdir()
(subdir / 'nested.txt').write_text('Nested file')
print()

# 4. éæ­·ç›®éŒ„ - iterdir()
print(f"éæ­·ç›®éŒ„ {new_dir}:")
for item in new_dir.iterdir():
    item_type = 'ç›®éŒ„' if item.is_dir() else 'æª”æ¡ˆ'
    print(f"  [{item_type}] {item.name}")
print()

# 5. åªåˆ—å‡ºæª”æ¡ˆ
print("åªåˆ—å‡ºæª”æ¡ˆ:")
for item in new_dir.iterdir():
    if item.is_file():
        print(f"  {item.name}")
print()

# 6. åˆªé™¤ç›®éŒ„ (å¿…é ˆç‚ºç©º)
print("æ¸…ç†:")
# å…ˆåˆªé™¤æ‰€æœ‰æª”æ¡ˆ
for item in new_dir.rglob('*'):  # rglob éè¿´æœå°‹
    if item.is_file():
        item.unlink()
        print(f"åˆªé™¤æª”æ¡ˆ: {item}")

# åˆªé™¤ç›®éŒ„ (å¾æœ€æ·±å±¤é–‹å§‹)
for item in sorted(new_dir.rglob('*'), reverse=True):
    if item.is_dir():
        item.rmdir()
        print(f"åˆªé™¤ç›®éŒ„: {item}")

new_dir.rmdir()
nested_dir.parent.parent.rmdir()  # level1
print("[æ¸…ç†å®Œæˆ]")
````

**é‡é»è§£æ**:
- `mkdir()`: å»ºç«‹ç›®éŒ„
  - `exist_ok=True`: ç›®éŒ„å·²å­˜åœ¨æ™‚ä¸å ±éŒ¯
  - `parents=True`: å»ºç«‹æ‰€æœ‰å¿…è¦çš„çˆ¶ç›®éŒ„
- `rmdir()`: åˆªé™¤ç©ºç›®éŒ„
- `iterdir()`: éæ­·ç›®éŒ„ (ä¸éè¿´)
- `rglob()`: éè¿´æœå°‹æ‰€æœ‰å­ç›®éŒ„

### ç¯„ä¾‹ 8: æª”æ¡ˆæ“ä½œ (rename, unlink, stat)

**å­¸ç¿’é‡é»**: é‡æ–°å‘½åã€åˆªé™¤ã€æŸ¥è©¢å±¬æ€§

````python
from pathlib import Path
from datetime import datetime

# å»ºç«‹æ¸¬è©¦æª”æ¡ˆ
test_file = Path('test_file.txt')
test_file.write_text('Hello, World!\nThis is a test file.')

print("æª”æ¡ˆæ“ä½œç¤ºä¾‹:")
print()

# 1. æŸ¥è©¢æª”æ¡ˆå±¬æ€§ - stat()
print("æª”æ¡ˆå±¬æ€§:")
stats = test_file.stat()
print(f"æª”æ¡ˆå¤§å°: {stats.st_size} bytes")
print(f"å»ºç«‹æ™‚é–“: {datetime.fromtimestamp(stats.st_ctime)}")
print(f"ä¿®æ”¹æ™‚é–“: {datetime.fromtimestamp(stats.st_mtime)}")
print(f"è¨ªå•æ™‚é–“: {datetime.fromtimestamp(stats.st_atime)}")
print()

# 2. è®€å–æª”æ¡ˆå…§å®¹
print("è®€å–å…§å®¹:")
content = test_file.read_text()
print(content)
print()

# 3. é‡æ–°å‘½åæª”æ¡ˆ
new_name = Path('renamed_file.txt')
test_file.rename(new_name)
print(f"æª”æ¡ˆå·²é‡æ–°å‘½å: {test_file.name} -> {new_name.name}")
print(f"æ–°æª”æ¡ˆå­˜åœ¨? {new_name.exists()}")
print()

# 4. è¤‡è£½æª”æ¡ˆ (éœ€è¦ä½¿ç”¨ shutil æˆ–æ‰‹å‹•è¤‡è£½)
copy_name = Path('copy_file.txt')
copy_name.write_text(new_name.read_text())
print(f"æª”æ¡ˆå·²è¤‡è£½: {copy_name.name}")
print()

# 5. åˆªé™¤æª”æ¡ˆ - unlink()
print("åˆªé™¤æª”æ¡ˆ:")
new_name.unlink()
print(f"{new_name.name} å·²åˆªé™¤")
print(f"æª”æ¡ˆé‚„å­˜åœ¨? {new_name.exists()}")

# 6. å®‰å…¨åˆªé™¤ (missing_ok=True, Python 3.8+)
copy_name.unlink(missing_ok=True)  # ä¸å­˜åœ¨ä¹Ÿä¸æœƒå ±éŒ¯
print(f"{copy_name.name} å·²åˆªé™¤")
print("[æ¸…ç†å®Œæˆ]")
````

**é‡é»è§£æ**:
- `stat()`: å–å¾—æª”æ¡ˆå±¬æ€§ (å¤§å°ã€æ™‚é–“ç­‰)
- `rename(target)`: é‡æ–°å‘½åæˆ–ç§»å‹•æª”æ¡ˆ

### ç¯„ä¾‹ 12: å¯¦æˆ° - æª”æ¡ˆçµ„ç¹”ç³»çµ±

**å­¸ç¿’é‡é»**: æ•´åˆæ‡‰ç”¨ï¼Œè¨­è¨ˆå¯¦ç”¨å·¥å…·

````python
from pathlib import Path
from datetime import datetime

def organize_files(source_dir, organize_by='extension'):
    """
    æ•´ç†ç›®éŒ„ä¸­çš„æª”æ¡ˆåˆ°åˆ†é¡è³‡æ–™å¤¾
    
    åƒæ•¸:
        source_dir: ä¾†æºç›®éŒ„è·¯å¾‘
        organize_by: åˆ†é¡æ–¹å¼ ('extension' æˆ– 'date')
    """
    source = Path(source_dir)
    
    if not source.exists() or not source.is_dir():
        print(f"éŒ¯èª¤: {source} ä¸æ˜¯æœ‰æ•ˆçš„ç›®éŒ„")
        return
    
    print(f"é–‹å§‹æ•´ç†ç›®éŒ„: {source}")
    print(f"åˆ†é¡æ–¹å¼: {organize_by}")
    print()
    
    moved_count = 0
    
    # éæ­·æ‰€æœ‰æª”æ¡ˆ (ä¸å«å­ç›®éŒ„)
    for file in source.glob('*'):
        if not file.is_file():
            continue
        
        # æ±ºå®šç›®æ¨™ç›®éŒ„
        if organize_by == 'extension':
            # æŒ‰å‰¯æª”ååˆ†é¡
            ext = file.suffix[1:] if file.suffix else 'å…¶ä»–'
            target_dir = source / ext
        elif organize_by == 'date':
            # æŒ‰ä¿®æ”¹æ—¥æœŸåˆ†é¡
            mtime = file.stat().st_mtime
            date = datetime.fromtimestamp(mtime).strftime('%Y-%m')
            target_dir = source / date
        else:
            print(f"ä¸æ”¯æ´çš„åˆ†é¡æ–¹å¼: {organize_by}")
            return
        
        # å»ºç«‹ç›®æ¨™ç›®éŒ„
        target_dir.mkdir(exist_ok=True)
        
        # ç§»å‹•æª”æ¡ˆ
        target_path = target_dir / file.name
        
        # è™•ç†åŒåæª”æ¡ˆ
        if target_path.exists():
            base = target_path.stem
            ext = target_path.suffix
            counter = 1
            while target_path.exists():
                target_path = target_dir / f"{base}_{counter}{ext}"
                counter += 1
        
        file.rename(target_path)
        print(f"ç§»å‹•: {file.name} -> {target_dir.name}/{target_path.name}")
        moved_count += 1
    
    print()
    print(f"æ•´ç†å®Œæˆï¼å…±ç§»å‹• {moved_count} å€‹æª”æ¡ˆ")

# æ¸¬è©¦
print("æ¸¬è©¦æª”æ¡ˆçµ„ç¹”ç³»çµ±:")
print("=" * 50)

# å»ºç«‹æ¸¬è©¦ç’°å¢ƒ
test_dir = Path('organize_test')
test_dir.mkdir(exist_ok=True)

# å»ºç«‹å„ç¨®æª”æ¡ˆ
test_files = [
    'document.txt', 'report.txt', 'data.csv',
    'image.png', 'photo.jpg', 'script.py',
    'notes.txt', 'archive.zip'
]

for file in test_files:
    (test_dir / file).write_text('test content')

# åŸ·è¡Œæ•´ç†
organize_files(test_dir, organize_by='extension')

print("\næ•´ç†å¾Œçš„ç›®éŒ„çµæ§‹:")
for item in sorted(test_dir.rglob('*')):
    if item.is_file():
        print(f"  {item.relative_to(test_dir)}")

# æ¸…ç†
for item in sorted(test_dir.rglob('*'), reverse=True):
    if item.is_file():
        item.unlink()
    elif item.is_dir():
        item.rmdir()
test_dir.rmdir()
print("\n[æ¸¬è©¦ç’°å¢ƒå·²æ¸…ç†]")
````
