### 習題 20：終極挑戦 - 貸款計算器 ⭐⭐⭐⭐⭐

計算貸款每月還款金額（等額本息）：

  ```

  ```
- 輸入：
  - 貸款金額：字串 `"1000000"`（100 萬）
  - 年利率：字串 `"0.02"`（2%）

  - 月利率 = 年利率 / 12

  - 總還款金額 = 每月還款 × 總期數
  - 總利息 = 總還款金額 - 貸款金額
- 輸出：
  ```
  貸款金額：1000000 元
  年利率：2.0%
  貸款年數：20 年（240 期）
  ─────────────────────

  總還款金額：1214155.20 元
  總利息：214155.20 元
  ```
