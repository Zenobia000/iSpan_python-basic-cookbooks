## ç¯„ä¾‹ 5ï¼šè¨ˆç®— BMIï¼ˆç¶œåˆæ‡‰ç”¨ï¼‰

### ğŸ“‹ å•é¡Œæè¿°
è¨ˆç®—èº«é«”è³ªé‡æŒ‡æ•¸ï¼ˆBMIï¼‰ä¸¦åˆ¤æ–·é«”é‡ç‹€æ…‹ï¼š
- å…¬å¼ï¼š`BMI = é«”é‡(kg) / èº«é«˜(m)Â²`
- è¼¸å…¥ï¼šé«”é‡ï¼ˆå…¬æ–¤ï¼‰ã€èº«é«˜ï¼ˆå…¬åˆ†ï¼‰
- è¼¸å‡ºï¼šBMI å€¼èˆ‡é«”é‡åˆ†é¡

BMI åˆ†é¡ï¼ˆå°ç£æ¨™æº–ï¼‰ï¼š

### ğŸ” åˆ†ææ€è·¯
1. è™•ç†è¼¸å…¥ï¼ˆå­—ä¸² â†’ æ•¸å­—ï¼‰

3. è¨ˆç®— BMI

### ğŸ’» é€æ­¥å¯¦ä½œ

````python
# Step 1: æ¨¡æ“¬ä½¿ç”¨è€…è¼¸å…¥ï¼ˆå­—ä¸²æ ¼å¼ï¼‰
weight_str = "70"       # é«”é‡ï¼ˆå…¬æ–¤ï¼‰
height_str = "175"      # èº«é«˜ï¼ˆå…¬åˆ†ï¼‰

print("ğŸ“‹ è¼¸å…¥è³‡æ–™")
print(f"é«”é‡ï¼š{weight_str} kg (å‹æ…‹: {type(weight_str).__name__})")
print(f"èº«é«˜ï¼š{height_str} cm (å‹æ…‹: {type(height_str).__name__})")
print()

# Step 2: å‹æ…‹è½‰æ›ï¼ˆstr â†’ floatï¼‰
weight_kg = float(weight_str)
height_cm = float(height_str)

# Step 3: å–®ä½è½‰æ›ï¼ˆå…¬åˆ† â†’ å…¬å°ºï¼‰
height_m = height_cm / 100
print(f"ğŸ”„ å–®ä½è½‰æ›ï¼š{height_cm} cm = {height_m} m")
print()

# Step 4: è¨ˆç®— BMI
bmi = weight_kg / (height_m ** 2)  # ** æ˜¯æ¬¡æ–¹é‹ç®—
print(f"ğŸ§® BMI è¨ˆç®—ï¼š{weight_kg} / ({height_m}Â²) = {bmi:.2f}")
print()

# Step 5: åˆ¤æ–·é«”é‡åˆ†é¡
if bmi < 18.5:
    category = "é«”é‡éè¼•"
elif bmi < 24:
    category = "æ­£å¸¸ç¯„åœ"
elif bmi < 27:
    category = "éé‡"
else:
    category = "è‚¥èƒ–"

# Step 6: é¡¯ç¤ºçµæœ
print("=" * 40)
print("ğŸ“Š BMI è©•ä¼°çµæœ")
print("=" * 40)
print(f"BMI å€¼ï¼š{bmi:.2f}")
print(f"é«”é‡åˆ†é¡ï¼š{category}")
print("=" * 40)
````
