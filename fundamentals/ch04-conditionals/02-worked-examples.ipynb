## ç¯„ä¾‹ 4ï¼šé–å¹´åˆ¤æ–· | Leap Year Detection

### å•é¡Œæè¿°
åˆ¤æ–·æŸå¹´æ˜¯å¦ç‚ºé–å¹´ã€‚é–å¹´è¦å‰‡ï¼š
1. èƒ½è¢« 4 æ•´é™¤ ä¸” ä¸èƒ½è¢« 100 æ•´é™¤ï¼Œæˆ–
2. èƒ½è¢« 400 æ•´é™¤

### åˆ†ææ€è·¯
é€™æ˜¯å…¸å‹çš„**è¤‡é›œé‚è¼¯é‹ç®—**ç¯„ä¾‹ï¼š
- è¦å‰‡ 1ï¼š`(year % 4 == 0) and (year % 100 != 0)`
- è¦å‰‡ 2ï¼š`year % 400 == 0`
- çµ„åˆï¼šè¦å‰‡ 1 **æˆ–** è¦å‰‡ 2

### å¯¦ä½œæ­¥é©Ÿ

````python
# æ­¥é©Ÿ 1ï¼šè¼¸å…¥è³‡æ–™
year = 2024

print("=== é–å¹´åˆ¤æ–·ç³»çµ± ===")
print(f"å¹´ä»½ï¼š{year}")

# æ­¥é©Ÿ 2ï¼šåˆ†è§£æ¢ä»¶åˆ¤æ–·
divisible_by_4 = (year % 4 == 0)
divisible_by_100 = (year % 100 == 0)
divisible_by_400 = (year % 400 == 0)

print(f"\nåˆ¤æ–·æ¢ä»¶ï¼š")
print(f"  èƒ½è¢« 4 æ•´é™¤ï¼š{divisible_by_4}")
print(f"  èƒ½è¢« 100 æ•´é™¤ï¼š{divisible_by_100}")
print(f"  èƒ½è¢« 400 æ•´é™¤ï¼š{divisible_by_400}")

# æ­¥é©Ÿ 3ï¼šæ‡‰ç”¨é–å¹´è¦å‰‡
is_leap = (divisible_by_4 and not divisible_by_100) or divisible_by_400

# æ­¥é©Ÿ 4ï¼šè¼¸å‡ºçµæœ
print(f"\nçµæœï¼š{year} å¹´{'æ˜¯' if is_leap else 'ä¸æ˜¯'}é–å¹´")

# æ­¥é©Ÿ 5ï¼šèªªæ˜åŸå› 
if is_leap:
    if divisible_by_400:
        print("åŸå› ï¼šèƒ½è¢« 400 æ•´é™¤")
    else:
        print("åŸå› ï¼šèƒ½è¢« 4 æ•´é™¤ä½†ä¸èƒ½è¢« 100 æ•´é™¤")
else:
    print("åŸå› ï¼šä¸ç¬¦åˆé–å¹´è¦å‰‡")
````

````python
# æ¸¬è©¦å¤šå€‹å¹´ä»½
test_years = [2000, 1900, 2024, 2023, 2100]

for year in test_years:
    is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
    status = "âœ… é–å¹´" if is_leap else "âŒ å¹³å¹´"
    print(f"{year}: {status}")
````

## ç¯„ä¾‹ 5ï¼šä¸‰è§’å½¢åˆ¤æ–· | Triangle Classification

### å•é¡Œæè¿°
çµ¦å®šä¸‰å€‹é‚Šé•·ï¼Œåˆ¤æ–·ï¼š
1. èƒ½å¦æ§‹æˆä¸‰è§’å½¢ï¼ˆä¸‰é‚Šé—œä¿‚ï¼‰
2. å¦‚æœå¯ä»¥ï¼Œæ˜¯ä»€éº¼é¡å‹çš„ä¸‰è§’å½¢ï¼š
   - ç­‰é‚Šä¸‰è§’å½¢ï¼ˆä¸‰é‚Šç›¸ç­‰ï¼‰
   - ç­‰è…°ä¸‰è§’å½¢ï¼ˆå…©é‚Šç›¸ç­‰ï¼‰
   - ç›´è§’ä¸‰è§’å½¢ï¼ˆç¬¦åˆç•¢æ°å®šç†ï¼‰
   - ä¸€èˆ¬ä¸‰è§’å½¢

### åˆ†ææ€è·¯
1. å…ˆæª¢æŸ¥ä¸‰é‚Šé—œä¿‚ï¼šä»»å…©é‚Šä¹‹å’Œ > ç¬¬ä¸‰é‚Š
2. å†åˆ¤æ–·ä¸‰è§’å½¢é¡å‹

### å¯¦ä½œæ­¥é©Ÿ

````python
# æ­¥é©Ÿ 1ï¼šè¼¸å…¥è³‡æ–™
a, b, c = 3, 4, 5

print("=== ä¸‰è§’å½¢åˆ¤æ–·ç³»çµ± ===")
print(f"ä¸‰é‚Šé•·ï¼ša={a}, b={b}, c={c}")

# æ­¥é©Ÿ 2ï¼šæª¢æŸ¥æ˜¯å¦èƒ½æ§‹æˆä¸‰è§’å½¢
can_form_triangle = (a + b > c) and (b + c > a) and (a + c > b)

if not can_form_triangle:
    print("\nâŒ ç„¡æ³•æ§‹æˆä¸‰è§’å½¢")
    print("åŸå› ï¼šä¸ç¬¦åˆä¸‰è§’å½¢ä¸‰é‚Šé—œä¿‚")
else:
    print("\nâœ… å¯ä»¥æ§‹æˆä¸‰è§’å½¢")
    
    # æ­¥é©Ÿ 3ï¼šåˆ¤æ–·ä¸‰è§’å½¢é¡å‹
    # æª¢æŸ¥ç­‰é‚Š
    if a == b == c:
        triangle_type = "ç­‰é‚Šä¸‰è§’å½¢"
        emoji = "ğŸ”º"
    # æª¢æŸ¥ç­‰è…°
    elif a == b or b == c or a == c:
        triangle_type = "ç­‰è…°ä¸‰è§’å½¢"
        emoji = "ğŸ“"
    else:
        triangle_type = "ä¸€èˆ¬ä¸‰è§’å½¢"
        emoji = "â–³"
    
    print(f"\n{emoji} é¡å‹ï¼š{triangle_type}")
    
    # æ­¥é©Ÿ 4ï¼šæª¢æŸ¥æ˜¯å¦ç‚ºç›´è§’ä¸‰è§’å½¢ï¼ˆç•¢æ°å®šç†ï¼‰
    sides = sorted([a, b, c])
    is_right = (sides[0]**2 + sides[1]**2 == sides[2]**2)
    
    if is_right:
        print("ğŸ“ ç‰¹æ®Šæ€§è³ªï¼šç›´è§’ä¸‰è§’å½¢")
        print(f"   ç¬¦åˆç•¢æ°å®šç†ï¼š{sides[0]}Â² + {sides[1]}Â² = {sides[2]}Â²")
````

````python
# æ¸¬è©¦ä¸åŒä¸‰è§’å½¢
test_cases = [
    (3, 3, 3, "ç­‰é‚Š"),
    (3, 3, 4, "ç­‰è…°"),
    (3, 4, 5, "ç›´è§’"),
    (2, 3, 4, "ä¸€èˆ¬"),
    (1, 2, 10, "ç„¡æ•ˆ"),
]

for a, b, c, expected in test_cases:
    can_form = (a + b > c) and (b + c > a) and (a + c > b)
    
    if not can_form:
        result = "ç„¡æ³•æ§‹æˆä¸‰è§’å½¢"
    else:
        if a == b == c:
            result = "ç­‰é‚Šä¸‰è§’å½¢"
        elif a == b or b == c or a == c:
            result = "ç­‰è…°ä¸‰è§’å½¢"
        else:
            result = "ä¸€èˆ¬ä¸‰è§’å½¢"
        
        sides = sorted([a, b, c])
        if sides[0]**2 + sides[1]**2 == sides[2]**2:
            result += " (ç›´è§’)"
    
    print(f"({a}, {b}, {c}) â†’ {result}")
````
