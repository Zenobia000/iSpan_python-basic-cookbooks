````python
import json
import os

# æ­¥é©Ÿ 1: å»ºç«‹é è¨­è¨­å®š
def create_default_config():
    """å»ºç«‹é è¨­è¨­å®šæª”"""
    config = {
        "app_info": {
            "name": "å­¸ç”Ÿç®¡ç†ç³»çµ±",
            "version": "1.0.0",
            "author": "é–‹ç™¼åœ˜éšŠ"
        },
        "database": {
            "host": "localhost",
            "port": 5432,
            "name": "student_db",
            "auto_backup": True
        },
        "ui": {
            "theme": "light",
            "language": "zh-TW",
            "font_size": 14
        },
        "features": {
            "enable_notifications": True,
            "max_upload_size_mb": 10,
            "session_timeout_minutes": 30
        }
    }
    
    # å„²å­˜åˆ°æª”æ¡ˆ
    with open('app_config.json', 'w', encoding='utf-8') as f:
        json.dump(config, f, ensure_ascii=False, indent=2)
    
    print("âœ“ é è¨­è¨­å®šæª”å·²å»ºç«‹")
    return config

# æ­¥é©Ÿ 2: è®€å–è¨­å®š
def load_config():
    """è¼‰å…¥è¨­å®šæª”"""
    if not os.path.exists('app_config.json'):
        print("âš  è¨­å®šæª”ä¸å­˜åœ¨ï¼Œå»ºç«‹é è¨­è¨­å®š")
        return create_default_config()
    
    with open('app_config.json', 'r', encoding='utf-8') as f:
        config = json.load(f)
    
    print("âœ“ è¨­å®šæª”å·²è¼‰å…¥")
    return config

# æ­¥é©Ÿ 3: æ›´æ–°è¨­å®š
def update_config(config, section, key, value):
    """æ›´æ–°ç‰¹å®šè¨­å®šé …ç›®"""
    if section in config and key in config[section]:
        old_value = config[section][key]
        config[section][key] = value
        print(f"âœ“ æ›´æ–° {section}.{key}: {old_value} â†’ {value}")
    else:
        print(f"âœ— æ‰¾ä¸åˆ°è¨­å®šé …ç›®: {section}.{key}")
    
    return config

# æ­¥é©Ÿ 4: å„²å­˜è¨­å®š
def save_config(config):
    """å„²å­˜è¨­å®šåˆ°æª”æ¡ˆ"""
    with open('app_config.json', 'w', encoding='utf-8') as f:
        json.dump(config, f, ensure_ascii=False, indent=2)
    
    print("âœ“ è¨­å®šå·²å„²å­˜")

# åŸ·è¡Œç¤ºç¯„
print("=== è¨­å®šæª”ç®¡ç†ç³»çµ± ===")
print()

# 1. å»ºç«‹æˆ–è¼‰å…¥è¨­å®š
config = load_config()
print()

# 2. é¡¯ç¤ºç•¶å‰è¨­å®š
print("ç•¶å‰è¨­å®š:")
print(json.dumps(config, ensure_ascii=False, indent=2))
print()

# 3. æ›´æ–°è¨­å®š
config = update_config(config, 'ui', 'theme', 'dark')
config = update_config(config, 'ui', 'font_size', 16)
config = update_config(config, 'features', 'enable_notifications', False)
print()

# 4. å„²å­˜è®Šæ›´
save_config(config)
print()

# 5. é©—è­‰ï¼šé‡æ–°è¼‰å…¥ä¸¦é¡¯ç¤º
print("é©—è­‰æ›´æ–°å¾Œçš„è¨­å®š:")
updated_config = load_config()
print(f"ä¸»é¡Œ: {updated_config['ui']['theme']}")
print(f"å­—é«”å¤§å°: {updated_config['ui']['font_size']}")
print(f"é€šçŸ¥: {updated_config['features']['enable_notifications']}")
````

````python
import json
from datetime import datetime

# æ¨¡æ“¬ API å›æ‡‰ï¼ˆçœŸå¯¦ API é€šå¸¸æœƒè¿”å›é€™æ¨£çš„ JSONï¼‰
api_response = '''
{
  "status": "success",
  "data": {
    "location": {
      "city": "å°åŒ—å¸‚",
      "district": "ä¿¡ç¾©å€",
      "coordinates": {"lat": 25.033, "lon": 121.565}
    },
    "current": {
      "temperature": 28.5,
      "feels_like": 30.2,
      "humidity": 75,
      "condition": "å¤šé›²æ™‚æ™´",
      "wind_speed": 12.3,
      "uv_index": 7
    },
    "forecast": [
      {
        "date": "2025-10-09",
        "day": "æ˜ŸæœŸä¸‰",
        "high": 30,
        "low": 24,
        "condition": "æ™´",
        "rain_chance": 20
      },
      {
        "date": "2025-10-10",
        "day": "æ˜ŸæœŸå››",
        "high": 29,
        "low": 23,
        "condition": "å¤šé›²",
        "rain_chance": 40
      },
      {
        "date": "2025-10-11",
        "day": "æ˜ŸæœŸäº”",
        "high": 27,
        "low": 22,
        "condition": "é™£é›¨",
        "rain_chance": 70
      }
    ]
  },
  "timestamp": "2025-10-08T14:30:00+08:00"
}
'''

# æ­¥é©Ÿ 1: è§£æ JSON
try:
    weather_data = json.loads(api_response)
    print("âœ“ API å›æ‡‰è§£ææˆåŠŸ")
except json.JSONDecodeError as e:
    print(f"âœ— JSON è§£æå¤±æ•—: {e}")
    weather_data = None

if weather_data:
    # æ­¥é©Ÿ 2: æå–è³‡æ–™
    status = weather_data['status']
    data = weather_data['data']
    
    location = data['location']
    current = data['current']
    forecast = data['forecast']
    
    # æ­¥é©Ÿ 3: é¡¯ç¤ºç•¶å‰å¤©æ°£
    print()
    print("=== ç•¶å‰å¤©æ°£è³‡è¨Š ===")
    print(f"åœ°é»: {location['city']} - {location['district']}")
    print(f"åº§æ¨™: ({location['coordinates']['lat']}, {location['coordinates']['lon']})")
    print(f"æº«åº¦: {current['temperature']}Â°C (é«”æ„Ÿ {current['feels_like']}Â°C)")
    print(f"å¤©æ°£: {current['condition']}")
    print(f"æ¿•åº¦: {current['humidity']}%")
    print(f"é¢¨é€Ÿ: {current['wind_speed']} km/h")
    print(f"ç´«å¤–ç·šæŒ‡æ•¸: {current['uv_index']}")
    
    # æ­¥é©Ÿ 4: é¡¯ç¤ºæœªä¾†é å ±
    print()
    print("=== æœªä¾†ä¸‰å¤©é å ± ===")
    print(f"{'æ—¥æœŸ':<12} {'æ˜ŸæœŸ':<8} {'é«˜æº«':<6} {'ä½æº«':<6} {'å¤©æ°£':<10} {'é™é›¨':<6}")
    print("-" * 60)
    
    for day in forecast:
        print(f"{day['date']:<12} {day['day']:<8} "
              f"{day['high']:>4}Â°C {day['low']:>4}Â°C "
              f"{day['condition']:<10} {day['rain_chance']:>3}%")
    
    # æ­¥é©Ÿ 5: è³‡æ–™åˆ†æ
    print()
    print("=== åˆ†æçµæœ ===")
    
    # è¨ˆç®—å¹³å‡é«˜æº«
    avg_high = sum(day['high'] for day in forecast) / len(forecast)
    print(f"å¹³å‡é«˜æº«: {avg_high:.1f}Â°C")
    
    # è¨ˆç®—å¹³å‡ä½æº«
    avg_low = sum(day['low'] for day in forecast) / len(forecast)
    print(f"å¹³å‡ä½æº«: {avg_low:.1f}Â°C")
    
    # æœ€é«˜é™é›¨æ©Ÿç‡
    max_rain = max(day['rain_chance'] for day in forecast)
    rainy_day = next(day for day in forecast if day['rain_chance'] == max_rain)
    print(f"æœ€é«˜é™é›¨æ©Ÿç‡: {max_rain}% ({rainy_day['day']})")
    
    # å»ºè­°
    print()
    print("=== å‡ºé–€å»ºè­° ===")
    if max_rain > 60:
        print("âœ“ è¨˜å¾—å¸¶å‚˜ï¼")
    if current['uv_index'] >= 7:
        print("âœ“ ç´«å¤–ç·šå¼·ï¼Œè«‹åšå¥½é˜²æ›¬")
    if current['temperature'] > 30:
        print("âœ“ å¤©æ°£ç‚ç†±ï¼Œå¤šè£œå……æ°´åˆ†")
    
    # æ­¥é©Ÿ 6: ç”Ÿæˆæ‘˜è¦ä¸¦å„²å­˜
    summary = {
        "åœ°é»": f"{location['city']}-{location['district']}",
        "æŸ¥è©¢æ™‚é–“": weather_data['timestamp'],
        "ç›®å‰æº«åº¦": current['temperature'],
        "ç›®å‰å¤©æ°£": current['condition'],
        "å¹³å‡é«˜æº«": round(avg_high, 1),
        "å¹³å‡ä½æº«": round(avg_low, 1),
        "éœ€è¦å¸¶å‚˜": max_rain > 60,
        "é å ±è©³æƒ…": forecast
    }
    
    with open('weather_summary.json', 'w', encoding='utf-8') as f:
        json.dump(summary, f, ensure_ascii=False, indent=2)
    
    print()
    print("âœ“ å¤©æ°£æ‘˜è¦å·²å„²å­˜åˆ° weather_summary.json")
````

## ç¯„ä¾‹ 3ï¼šä½¿ç”¨è€…è³‡æ–™å„²å­˜ç³»çµ±

### ğŸ“‹ å•é¡Œæè¿°

å¯¦ä½œä¸€å€‹ç°¡æ˜“çš„ä½¿ç”¨è€…è³‡æ–™ç®¡ç†ç³»çµ±ï¼Œæ”¯æ´ CRUD æ“ä½œï¼š
1. **Create**: æ–°å¢ä½¿ç”¨è€…
2. **Read**: æŸ¥è©¢ä½¿ç”¨è€…
3. **Update**: æ›´æ–°ä½¿ç”¨è€…è³‡æ–™
4. **Delete**: åˆªé™¤ä½¿ç”¨è€…

**é›£åº¦**ï¼šä¸­ç´š

### ğŸ” åˆ†ææ€è·¯

1. **è³‡æ–™çµæ§‹**ï¼šä½¿ç”¨åˆ—è¡¨å„²å­˜å¤šå€‹ä½¿ç”¨è€…å­—å…¸
2. **æŒä¹…åŒ–**ï¼šJSON æª”æ¡ˆä½œç‚ºç°¡æ˜“è³‡æ–™åº«
3. **å”¯ä¸€è­˜åˆ¥**ï¼šä½¿ç”¨ user_id ä½œç‚ºä¸»éµ
4. **æ“ä½œå°è£**ï¼šæ¯å€‹æ“ä½œç¨ç«‹ç‚ºå‡½å¼

### ğŸ’» é€æ­¥å¯¦ä½œ

````python
import json
import os

# è³‡æ–™åº«æª”æ¡ˆ
DB_FILE = 'users_db.json'

# === è¼”åŠ©å‡½å¼ ===

def load_users():
    """è¼‰å…¥ä½¿ç”¨è€…è³‡æ–™"""
    if not os.path.exists(DB_FILE):
        return []
    
    with open(DB_FILE, 'r', encoding='utf-8') as f:
        return json.load(f)

def save_users(users):
    """å„²å­˜ä½¿ç”¨è€…è³‡æ–™"""
    with open(DB_FILE, 'w', encoding='utf-8') as f:
        json.dump(users, f, ensure_ascii=False, indent=2)

def generate_user_id(users):
    """ç”Ÿæˆæ–°çš„ä½¿ç”¨è€… ID"""
    if not users:
        return 1
    return max(user['user_id'] for user in users) + 1

# === CRUD æ“ä½œ ===

def create_user(name, email, age):
    """æ–°å¢ä½¿ç”¨è€…"""
    users = load_users()
    
    # æª¢æŸ¥ email æ˜¯å¦å·²å­˜åœ¨
    if any(user['email'] == email for user in users):
        print(f"âœ— Email {email} å·²å­˜åœ¨")
        return None
    
    # å»ºç«‹æ–°ä½¿ç”¨è€…
    new_user = {
        'user_id': generate_user_id(users),
        'name': name,
        'email': email,
        'age': age,
        'active': True
    }
    
    users.append(new_user)
    save_users(users)
    
    print(f"âœ“ ä½¿ç”¨è€… {name} å·²å»ºç«‹ (ID: {new_user['user_id']})")
    return new_user

def read_user(user_id):
    """æŸ¥è©¢ä½¿ç”¨è€…"""
    users = load_users()
    
    for user in users:
        if user['user_id'] == user_id:
            return user
    
    print(f"âœ— æ‰¾ä¸åˆ°ä½¿ç”¨è€… ID: {user_id}")
    return None

def update_user(user_id, **kwargs):
    """æ›´æ–°ä½¿ç”¨è€…è³‡æ–™"""
    users = load_users()
    
    for user in users:
        if user['user_id'] == user_id:
            # æ›´æ–°æä¾›çš„æ¬„ä½
            for key, value in kwargs.items():
                if key in user:
                    user[key] = value
            
            save_users(users)
            print(f"âœ“ ä½¿ç”¨è€… ID {user_id} å·²æ›´æ–°")
            return user
    
    print(f"âœ— æ‰¾ä¸åˆ°ä½¿ç”¨è€… ID: {user_id}")
    return None

def delete_user(user_id):
    """åˆªé™¤ä½¿ç”¨è€…"""
    users = load_users()
    
    for i, user in enumerate(users):
        if user['user_id'] == user_id:
            deleted_user = users.pop(i)
            save_users(users)
            print(f"âœ“ ä½¿ç”¨è€… {deleted_user['name']} å·²åˆªé™¤")
            return deleted_user
    
    print(f"âœ— æ‰¾ä¸åˆ°ä½¿ç”¨è€… ID: {user_id}")
    return None

def list_all_users():
    """åˆ—å‡ºæ‰€æœ‰ä½¿ç”¨è€…"""
    users = load_users()
    
    if not users:
        print("ç›®å‰æ²’æœ‰ä½¿ç”¨è€…")
        return
    
    print(f"\n{'ID':<5} {'å§“å':<15} {'Email':<25} {'å¹´é½¡':<5} {'ç‹€æ…‹':<5}")
    print("-" * 60)
    
    for user in users:
        status = "å•Ÿç”¨" if user['active'] else "åœç”¨"
        print(f"{user['user_id']:<5} {user['name']:<15} "
              f"{user['email']:<25} {user['age']:<5} {status:<5}")

# === æ¸¬è©¦ CRUD æ“ä½œ ===

print("=== ä½¿ç”¨è€…è³‡æ–™ç®¡ç†ç³»çµ± ===")
print()

# 1. Create - æ–°å¢ä½¿ç”¨è€…
print("[1] æ–°å¢ä½¿ç”¨è€…")
create_user("ç‹å°æ˜", "ming@example.com", 25)
create_user("æç¾è¯", "mei@example.com", 28)
create_user("å¼µå¤§å‰", "wei@example.com", 32)
print()

# 2. Read - æŸ¥è©¢ä½¿ç”¨è€…
print("[2] æŸ¥è©¢ä½¿ç”¨è€…")
user = read_user(1)
if user:
    print(f"æ‰¾åˆ°ä½¿ç”¨è€…: {user['name']} ({user['email']})")
print()

# 3. List - åˆ—å‡ºæ‰€æœ‰ä½¿ç”¨è€…
print("[3] åˆ—å‡ºæ‰€æœ‰ä½¿ç”¨è€…")
list_all_users()
print()

# 4. Update - æ›´æ–°ä½¿ç”¨è€…
print("[4] æ›´æ–°ä½¿ç”¨è€…è³‡æ–™")
update_user(2, age=29, active=False)
list_all_users()
print()

# 5. Delete - åˆªé™¤ä½¿ç”¨è€…
print("[5] åˆªé™¤ä½¿ç”¨è€…")
delete_user(3)
list_all_users()
print()

print("âœ“ æ‰€æœ‰è³‡æ–™å·²å„²å­˜åˆ°", DB_FILE)
````

## ç¯„ä¾‹ 4ï¼šCSV è½‰ JSON å·¥å…·

### ğŸ“‹ å•é¡Œæè¿°

å»ºç«‹ä¸€å€‹å·¥å…·ï¼Œå°‡ CSV æ ¼å¼çš„å­¸ç”Ÿæˆç¸¾è³‡æ–™è½‰æ›ç‚º JSON æ ¼å¼ï¼š

1. **CSV è§£æ**ï¼šç¬¬ä¸€è¡Œç‚ºæ¨™é¡Œï¼Œå…¶é¤˜ç‚ºè³‡æ–™

````python
import json

# æ­¥é©Ÿ 1: å»ºç«‹ç¯„ä¾‹ CSV è³‡æ–™
csv_content = """student_id,name,math,english,science
1001,ç‹å°æ˜,85,78,92
1002,æç¾è¯,92,88,95
1003,å¼µå¤§å‰,78,85,80
1004,é™³æ·‘èŠ¬,88,92,87
1005,æ—å¿—è±ª,95,90,93"""

# å¯«å…¥ CSV æª”æ¡ˆ
with open('students.csv', 'w', encoding='utf-8') as f:
    f.write(csv_content)

print("âœ“ CSV æª”æ¡ˆå·²å»ºç«‹")
print()

# æ­¥é©Ÿ 2: è®€å–ä¸¦è§£æ CSV
def csv_to_json(csv_file, json_file):
    """å°‡ CSV æª”æ¡ˆè½‰æ›ç‚º JSON"""
    
    # è®€å– CSV æª”æ¡ˆ
    with open(csv_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    # è§£ææ¨™é¡Œè¡Œ
    headers = lines[0].strip().split(',')
    print(f"æ¬„ä½: {headers}")
    print()
    
    # è§£æè³‡æ–™è¡Œ
    students = []
    for line in lines[1:]:
        values = line.strip().split(',')
        
        # å»ºç«‹å­—å…¸
        student = {}
        for i, header in enumerate(headers):
            value = values[i]
            
            # å˜—è©¦è½‰æ›ç‚ºæ•¸å­—
            try:
                # å…ˆå˜—è©¦æ•´æ•¸
                student[header] = int(value)
            except ValueError:
                try:
                    # å†å˜—è©¦æµ®é»æ•¸
                    student[header] = float(value)
                except ValueError:
                    # ä¿æŒå­—ä¸²
                    student[header] = value
        
        # è¨ˆç®—å¹³å‡åˆ†æ•¸
        scores = [student['math'], student['english'], student['science']]
        student['average'] = round(sum(scores) / len(scores), 2)
        
        students.append(student)
    
    print(f"âœ“ è§£æäº† {len(students)} ç­†å­¸ç”Ÿè³‡æ–™")
    
    # å„²å­˜ç‚º JSON
    with open(json_file, 'w', encoding='utf-8') as f:
        json.dump(students, f, ensure_ascii=False, indent=2)
    
    print(f"âœ“ å·²å„²å­˜åˆ° {json_file}")
    
    return students

# æ­¥é©Ÿ 3: åŸ·è¡Œè½‰æ›
print("=== CSV è½‰ JSON å·¥å…· ===")
print()

students = csv_to_json('students.csv', 'students.json')
print()

# æ­¥é©Ÿ 4: é¡¯ç¤ºè½‰æ›çµæœ
print("è½‰æ›çµæœé è¦½:")
print(json.dumps(students[:2], ensure_ascii=False, indent=2))
print("...")
print()

# æ­¥é©Ÿ 5: çµ±è¨ˆåˆ†æ
print("=== çµ±è¨ˆåˆ†æ ===")
print(f"ç¸½å­¸ç”Ÿæ•¸: {len(students)}")

avg_math = sum(s['math'] for s in students) / len(students)
avg_english = sum(s['english'] for s in students) / len(students)
avg_science = sum(s['science'] for s in students) / len(students)

print(f"æ•¸å­¸å¹³å‡: {avg_math:.2f}")
print(f"è‹±æ–‡å¹³å‡: {avg_english:.2f}")
print(f"è‡ªç„¶å¹³å‡: {avg_science:.2f}")
print()

# æ‰¾å‡ºæœ€é«˜å¹³å‡åˆ†æ•¸çš„å­¸ç”Ÿ
top_student = max(students, key=lambda s: s['average'])
print(f"æœ€é«˜å¹³å‡åˆ†: {top_student['name']} ({top_student['average']}åˆ†)")
````

3. **å‹æ…‹è™•ç†**ï¼šé‡å°æ¯ç¨®å‹æ…‹æä¾›è½‰æ›é‚è¼¯
4. **ååºåˆ—åŒ–**ï¼šè¨­è¨ˆå°æ‡‰çš„è§£æé‚è¼¯

### ğŸ’» é€æ­¥å¯¦ä½œ

````python
import json
from datetime import datetime, date
from decimal import Decimal

# æ­¥é©Ÿ 1: è‡ªè¨‚ JSON ç·¨ç¢¼å™¨
class CustomJSONEncoder(json.JSONEncoder):
    """è‡ªè¨‚ JSON ç·¨ç¢¼å™¨ï¼Œè™•ç†ç‰¹æ®Šå‹æ…‹"""
    
    def default(self, obj):
        # è™•ç† datetime
        if isinstance(obj, datetime):
            return {
                '__type__': 'datetime',
                'value': obj.isoformat()
            }
        
        # è™•ç† date
        if isinstance(obj, date):
            return {
                '__type__': 'date',
                'value': obj.isoformat()
            }
        
        # è™•ç† set
        if isinstance(obj, set):
            return {
                '__type__': 'set',
                'value': list(obj)
            }
        
        # è™•ç† Decimal
        if isinstance(obj, Decimal):
            return {
                '__type__': 'decimal',
                'value': str(obj)
            }
        
        # å…¶ä»–äº¤çµ¦çˆ¶é¡è™•ç†
        return super().default(obj)

# æ­¥é©Ÿ 2: è‡ªè¨‚ JSON è§£ç¢¼å™¨
def custom_decoder(dct):
    """è‡ªè¨‚ JSON è§£ç¢¼å™¨ï¼Œé‚„åŸç‰¹æ®Šå‹æ…‹"""
    
    if '__type__' not in dct:
        return dct
    
    obj_type = dct['__type__']
    value = dct['value']
    
    # é‚„åŸ datetime
    if obj_type == 'datetime':
        return datetime.fromisoformat(value)
    
    # é‚„åŸ date
    if obj_type == 'date':
        return date.fromisoformat(value)
    
    # é‚„åŸ set
    if obj_type == 'set':
        return set(value)
    
    # é‚„åŸ Decimal
    if obj_type == 'decimal':
        return Decimal(value)
    
    return dct

# æ­¥é©Ÿ 3: æ¸¬è©¦è³‡æ–™
order = {
    'order_id': 'ORD-2025-001',
    'customer': 'ç‹å°æ˜',
    'order_date': datetime(2025, 10, 8, 14, 30, 0),
    'delivery_date': date(2025, 10, 10),
    'total_amount': Decimal('1299.99'),
    'items': [
        {'name': 'Python æ›¸ç±', 'price': Decimal('599.00'), 'quantity': 2},
        {'name': 'USB éš¨èº«ç¢Ÿ', 'price': Decimal('350.99'), 'quantity': 1}
    ],
    'tags': {'é›»å­ç”¢å“', 'æ›¸ç±', 'ä¿ƒéŠ·'},  # set
    'status': 'pending'
}

print("=== è¤‡é›œç‰©ä»¶åºåˆ—åŒ– ===")
print()
print("åŸå§‹è³‡æ–™:")
print(f"è¨‚å–® ID: {order['order_id']}")
print(f"è¨‚å–®æ—¥æœŸ: {order['order_date']} (å‹æ…‹: {type(order['order_date']).__name__})")
print(f"ç¸½é‡‘é¡: {order['total_amount']} (å‹æ…‹: {type(order['total_amount']).__name__})")
print(f"æ¨™ç±¤: {order['tags']} (å‹æ…‹: {type(order['tags']).__name__})")
print()

# æ­¥é©Ÿ 4: åºåˆ—åŒ–ï¼ˆä½¿ç”¨è‡ªè¨‚ç·¨ç¢¼å™¨ï¼‰
json_string = json.dumps(order, cls=CustomJSONEncoder, ensure_ascii=False, indent=2)

print("åºåˆ—åŒ–çµæœ:")
print(json_string)
print()

# æ­¥é©Ÿ 5: å„²å­˜åˆ°æª”æ¡ˆ
with open('order.json', 'w', encoding='utf-8') as f:
    json.dump(order, f, cls=CustomJSONEncoder, ensure_ascii=False, indent=2)

print("âœ“ å·²å„²å­˜åˆ° order.json")
print()

# æ­¥é©Ÿ 6: ååºåˆ—åŒ–ï¼ˆä½¿ç”¨è‡ªè¨‚è§£ç¢¼å™¨ï¼‰
with open('order.json', 'r', encoding='utf-8') as f:
    loaded_order = json.load(f, object_hook=custom_decoder)

print("ååºåˆ—åŒ–çµæœ:")
print(f"è¨‚å–® ID: {loaded_order['order_id']}")
print(f"è¨‚å–®æ—¥æœŸ: {loaded_order['order_date']} (å‹æ…‹: {type(loaded_order['order_date']).__name__})")
print(f"ç¸½é‡‘é¡: {loaded_order['total_amount']} (å‹æ…‹: {type(loaded_order['total_amount']).__name__})")
print(f"æ¨™ç±¤: {loaded_order['tags']} (å‹æ…‹: {type(loaded_order['tags']).__name__})")
print()

# æ­¥é©Ÿ 7: é©—è­‰å‹æ…‹é‚„åŸ
print("=== å‹æ…‹é©—è­‰ ===")
print(f"datetime é‚„åŸæ­£ç¢º: {isinstance(loaded_order['order_date'], datetime)}")
print(f"date é‚„åŸæ­£ç¢º: {isinstance(loaded_order['delivery_date'], date)}")
print(f"Decimal é‚„åŸæ­£ç¢º: {isinstance(loaded_order['total_amount'], Decimal)}")
print(f"set é‚„åŸæ­£ç¢º: {isinstance(loaded_order['tags'], set)}")
print()

# é©—è­‰è³‡æ–™å®Œæ•´æ€§
print("âœ“ æ‰€æœ‰å‹æ…‹æˆåŠŸé‚„åŸï¼")
````
