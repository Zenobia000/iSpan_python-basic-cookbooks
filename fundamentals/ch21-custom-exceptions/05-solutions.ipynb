---## 基礎題 1: NegativeNumberError（10 分）### 解題思路1. 繼承 `ValueError`（因為是值的錯誤）2. 儲存 `value` 屬性3. 在 `sqrt` 函式中檢查並拋出例外### 完整解答

````python
# 基礎題 1 解答class NegativeNumberError(ValueError):    """負數錯誤"""    def __init__(self, value):        self.value = value        super().__init__(f"負數不被允許：{value}")def sqrt(x):    """計算平方根"""    if x < 0:        raise NegativeNumberError(x)    return x ** 0.5# 測試print(sqrt(16))  # 4.0try:    sqrt(-9)except NegativeNumberError as e:    print(f"錯誤：{e}, 值={e.value}")
````

---## 基礎題 2: InvalidCredentialsError（10 分）### 解題思路1. 根據 `attempt_count` 計算 `is_locked`2. 根據鎖定狀態產生不同訊息### 完整解答

````python
# 基礎題 2 解答class InvalidCredentialsError(Exception):    """登入驗證失敗"""    def __init__(self, username, attempt_count):        self.username = username        self.attempt_count = attempt_count        self.is_locked = (attempt_count >= 3)        if self.is_locked:            message = f"使用者 {username} 已被鎖定（失敗 {attempt_count} 次）"        else:            message = f"使用者 {username} 登入失敗（第 {attempt_count} 次）"        super().__init__(message)# 測試try:    raise InvalidCredentialsError("alice", 1)except InvalidCredentialsError as e:    print(f"{e}, 鎖定={e.is_locked}")try:    raise InvalidCredentialsError("alice", 3)except InvalidCredentialsError as e:    print(f"{e}, 鎖定={e.is_locked}")
````
