````python
# src/storage.py ç¯„æœ¬
"""è³‡æ–™å„²å­˜æ¨¡çµ„

è² è²¬æª”æ¡ˆè®€å¯«æ“ä½œèˆ‡è³‡æ–™åºåˆ—åŒ–ã€‚
"""

import json
from pathlib import Path
from typing import List, Dict, Any


# ========== JSON æª”æ¡ˆæ“ä½œ ==========

def load_json(file_path: Path) -> List[Dict[str, Any]]:
    """å¾ JSON æª”æ¡ˆè¼‰å…¥è³‡æ–™
    
    Args:
        file_path: æª”æ¡ˆè·¯å¾‘
    
    Returns:
        è³‡æ–™åˆ—è¡¨
    
    Raises:
        FileNotFoundError: æª”æ¡ˆä¸å­˜åœ¨æ™‚
        json.JSONDecodeError: JSON æ ¼å¼éŒ¯èª¤æ™‚
    """
    if not file_path.exists():
        return []
    
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            return json.load(f)
    except json.JSONDecodeError as e:
        print(f"JSON æ ¼å¼éŒ¯èª¤: {e}")
        return []


def save_json(data: List[Dict[str, Any]], file_path: Path) -> bool:
    """å„²å­˜è³‡æ–™åˆ° JSON æª”æ¡ˆ
    
    Args:
        data: è¦å„²å­˜çš„è³‡æ–™
        file_path: æª”æ¡ˆè·¯å¾‘
    
    Returns:
        æ˜¯å¦æˆåŠŸå„²å­˜
    """
    try:
        # ç¢ºä¿ç›®éŒ„å­˜åœ¨
        file_path.parent.mkdir(parents=True, exist_ok=True)
        
        with open(file_path, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=2)
        
        return True
    except Exception as e:
        print(f"å„²å­˜å¤±æ•—: {e}")
        return False


# ========== CSV æª”æ¡ˆæ“ä½œ (å¦‚éœ€è¦) ==========

# import csv
# 
# def load_csv(file_path: Path) -> List[Dict[str, str]]:
#     """å¾ CSV æª”æ¡ˆè¼‰å…¥è³‡æ–™"""
#     pass
# 
# def save_csv(data: List[Dict[str, str]], file_path: Path) -> bool:
#     """å„²å­˜è³‡æ–™åˆ° CSV æª”æ¡ˆ"""
#     pass
````
