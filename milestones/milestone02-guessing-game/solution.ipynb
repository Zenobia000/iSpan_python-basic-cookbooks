````python
"""åŸºæœ¬ç‰ˆæœ¬ï¼šçµ‚æ¥µå¯†ç¢¼éŠæˆ²"""

import random

# éŠæˆ²å¸¸æ•¸è¨­å®š
MAX_ATTEMPTS = 7
MIN_NUM = 1
MAX_NUM = 100

# ç”¢ç”Ÿéš¨æ©Ÿç­”æ¡ˆ
secret_number = random.randint(MIN_NUM, MAX_NUM)
attempts = 0

# éŠæˆ²é–‹å§‹è¨Šæ¯
print("=" * 50)
print("       ğŸ® çµ‚æ¥µå¯†ç¢¼éŠæˆ² Ultimate Password       ")
print("=" * 50)
print(f"\nç¯„åœï¼š{MIN_NUM}-{MAX_NUM}ï¼Œä½ æœ‰ {MAX_ATTEMPTS} æ¬¡æ©Ÿæœƒï¼")
print("è«‹çŒœå‡ºé›»è…¦æƒ³çš„æ•¸å­—ï¼\n")

# éŠæˆ²ä¸»è¿´åœˆ
while attempts < MAX_ATTEMPTS:
    try:
        # å¢åŠ çŒœæ¸¬æ¬¡æ•¸
        attempts += 1
        
        # é¡¯ç¤ºç•¶å‰ç‹€æ…‹
        remaining = MAX_ATTEMPTS - attempts
        print(f"ç¬¬ {attempts} æ¬¡çŒœæ¸¬ï¼ˆå‰©é¤˜ {remaining} æ¬¡æ©Ÿæœƒï¼‰")
        
        # å–å¾—ç©å®¶è¼¸å…¥
        guess = int(input(f"è«‹è¼¸å…¥ä½ çš„çŒœæ¸¬ ({MIN_NUM}-{MAX_NUM}): "))
        
        # é©—è­‰ç¯„åœ
        if guess < MIN_NUM or guess > MAX_NUM:
            print(f"âŒ è«‹è¼¸å…¥ {MIN_NUM}-{MAX_NUM} ä¹‹é–“çš„æ•¸å­—ï¼\n")
            attempts -= 1  # ä¸è¨ˆå…¥æ­¤æ¬¡çŒœæ¸¬
            continue
        
        # åˆ¤æ–·çŒœæ¸¬çµæœ
        if guess == secret_number:
            # å‹åˆ©
            print("\n" + "=" * 50)
            print("ğŸ‰ æ­å–œç­”å°äº†ï¼")
            print(f"ä½ ç¸½å…±çŒœäº† {attempts} æ¬¡ã€‚")
            
            # æ ¹æ“šæ¬¡æ•¸çµ¦äºˆè©•åƒ¹
            if attempts <= 3:
                print("ğŸ’¡ å¤ªå²å®³äº†ï¼ä½ æ˜¯çŒœæ•¸å­—å¤§å¸«ï¼")
            elif attempts <= 5:
                print("ğŸ‘ è¡¨ç¾ä¸éŒ¯ï¼")
            else:
                print("âœ… æˆåŠŸå®Œæˆï¼")
            
            print("=" * 50)
            break
            
        elif guess > secret_number:
            print("ğŸ“‰ å¤ªå¤§äº†ï¼\n")
        else:
            print("ğŸ“ˆ å¤ªå°äº†ï¼\n")
    
    except ValueError:
        # è™•ç†éæ•¸å­—è¼¸å…¥
        print("âŒ è«‹è¼¸å…¥æœ‰æ•ˆçš„æ•¸å­—ï¼\n")
        attempts -= 1  # ä¸è¨ˆå…¥æ­¤æ¬¡çŒœæ¸¬
        continue

# éŠæˆ²çµæŸåˆ¤æ–·
if attempts >= MAX_ATTEMPTS and guess != secret_number:
    # å¤±æ•—
    print("\n" + "=" * 50)
    print(f"ğŸ˜¢ å¾ˆéºæ†¾ï¼Œä½ å·²ç”¨å®Œ {MAX_ATTEMPTS} æ¬¡æ©Ÿæœƒï¼")
    print(f"æ­£ç¢ºç­”æ¡ˆæ˜¯ï¼š{secret_number}")
    print("ğŸ’ª åˆ¥æ°£é¤’ï¼Œå†è©¦ä¸€æ¬¡å§ï¼")
    print("=" * 50)

print("\næ„Ÿè¬éŠç©ï¼")
````

### æ–°å¢åŠŸèƒ½
- âœ… é›£åº¦é¸æ“‡ï¼ˆç°¡å–®/ä¸­ç­‰/å›°é›£ï¼‰
- âœ… çŒœæ¸¬æ­·å²è¨˜éŒ„
- âœ… é‡æ–°éŠæˆ²åŠŸèƒ½
- âœ… éŠæˆ²çµ±è¨ˆï¼ˆç¸½å±€æ•¸ã€å‹ç‡ã€æœ€ä½³ç´€éŒ„ï¼‰
- âœ… æ¨¡çµ„åŒ–è¨­è¨ˆï¼ˆå‡½å¼æ‹†åˆ†ï¼‰

````python
"""é€²éšç‰ˆæœ¬ï¼šçµ‚æ¥µå¯†ç¢¼éŠæˆ²ï¼ˆå®Œæ•´åŠŸèƒ½ï¼‰"""

import random

def choose_difficulty():
    """
    é¸æ“‡éŠæˆ²é›£åº¦
    
    å›å‚³ï¼š
        tuple: (min_num, max_num, max_attempts)
    """
    print("\n" + "=" * 40)
    print("           é¸æ“‡é›£åº¦ Select Difficulty")
    print("=" * 40)
    print("1. ğŸŸ¢ ç°¡å–® Easy    (1-50,   10 æ¬¡æ©Ÿæœƒ)")
    print("2. ğŸŸ¡ ä¸­ç­‰ Medium  (1-100,   7 æ¬¡æ©Ÿæœƒ)")
    print("3. ğŸ”´ å›°é›£ Hard    (1-200,   6 æ¬¡æ©Ÿæœƒ)")
    print("=" * 40)
    
    while True:
        try:
            choice = int(input("\nè«‹é¸æ“‡é›£åº¦ (1-3): "))
            
            if choice == 1:
                print("\nâœ… å·²é¸æ“‡ï¼šç°¡å–®æ¨¡å¼")
                return (1, 50, 10)
            elif choice == 2:
                print("\nâœ… å·²é¸æ“‡ï¼šä¸­ç­‰æ¨¡å¼")
                return (1, 100, 7)
            elif choice == 3:
                print("\nâœ… å·²é¸æ“‡ï¼šå›°é›£æ¨¡å¼")
                return (1, 200, 6)
            else:
                print("âŒ è«‹è¼¸å…¥ 1-3 ä¹‹é–“çš„æ•¸å­—ï¼")
        except ValueError:
            print("âŒ è«‹è¼¸å…¥æœ‰æ•ˆçš„æ•¸å­—ï¼")


def get_valid_guess(min_num, max_num, attempt_num, max_attempts):
    """
    å–å¾—æœ‰æ•ˆçš„ç©å®¶çŒœæ¸¬ï¼ˆåŒ…å«è¼¸å…¥é©—è­‰ï¼‰
    
    åƒæ•¸ï¼š
        min_num: æœ€å°å€¼
        max_num: æœ€å¤§å€¼
        attempt_num: ç•¶å‰çŒœæ¸¬æ¬¡æ•¸
        max_attempts: æœ€å¤§çŒœæ¸¬æ¬¡æ•¸
    
    å›å‚³ï¼š
        int: æœ‰æ•ˆçš„çŒœæ¸¬æ•¸å­—
    """
    while True:
        try:
            remaining = max_attempts - attempt_num
            print(f"\nğŸ“Œ ç¬¬ {attempt_num} æ¬¡çŒœæ¸¬ï¼ˆå‰©é¤˜ {remaining} æ¬¡æ©Ÿæœƒï¼‰")
            guess = int(input(f"è«‹è¼¸å…¥ä½ çš„çŒœæ¸¬ ({min_num}-{max_num}): "))
            
            # æª¢æŸ¥ç¯„åœ
            if guess < min_num or guess > max_num:
                print(f"âŒ è«‹è¼¸å…¥ {min_num}-{max_num} ä¹‹é–“çš„æ•¸å­—ï¼")
                continue
            
            return guess
            
        except ValueError:
            print("âŒ è«‹è¼¸å…¥æœ‰æ•ˆçš„æ•¸å­—ï¼")


def check_guess(guess, secret):
    """
    æª¢æŸ¥çŒœæ¸¬çµæœ
    
    åƒæ•¸ï¼š
        guess: ç©å®¶çŒœæ¸¬
        secret: æ­£ç¢ºç­”æ¡ˆ
    
    å›å‚³ï¼š
        str: 'correct', 'too_high', 'too_low'
    """
    if guess == secret:
        return 'correct'
    elif guess > secret:
        return 'too_high'
    else:
        return 'too_low'


def show_guess_history(history):
    """
    é¡¯ç¤ºçŒœæ¸¬æ­·å²
    
    åƒæ•¸ï¼š
        history: çŒœæ¸¬æ­·å²ä¸²åˆ—
    """
    if not history:
        return
    
    print(f"\nğŸ“Š çŒœæ¸¬æ­·å²ï¼š{history}")
    if len(history) > 0:
        print(f"   ç¯„åœï¼š{min(history)} - {max(history)}")


def show_victory_message(attempts):
    """é¡¯ç¤ºå‹åˆ©è¨Šæ¯"""
    print("\n" + "=" * 50)
    print("ğŸ‰ğŸ‰ğŸ‰ æ­å–œç­”å°äº†ï¼You Win! ğŸ‰ğŸ‰ğŸ‰")
    print("=" * 50)
    print(f"âœ¨ ä½ ç¸½å…±çŒœäº† {attempts} æ¬¡ã€‚")
    
    # æ ¹æ“šæ¬¡æ•¸çµ¦äºˆè©•åƒ¹
    if attempts <= 3:
        print("ğŸ† å¤ªå²å®³äº†ï¼ä½ æ˜¯çŒœæ•¸å­—å¤§å¸«ï¼")
    elif attempts <= 5:
        print("ğŸ‘ è¡¨ç¾ä¸éŒ¯ï¼")
    elif attempts <= 7:
        print("âœ… æˆåŠŸå®Œæˆï¼")
    else:
        print("ğŸ’ª å …æŒåˆ°åº•ï¼")
    
    print("=" * 50)


def show_defeat_message(secret, max_attempts):
    """é¡¯ç¤ºå¤±æ•—è¨Šæ¯"""
    print("\n" + "=" * 50)
    print("ğŸ˜¢ å¾ˆéºæ†¾ï¼ŒGame Over!")
    print("=" * 50)
    print(f"ä½ å·²ç”¨å®Œ {max_attempts} æ¬¡æ©Ÿæœƒï¼")
    print(f"âœ¨ æ­£ç¢ºç­”æ¡ˆæ˜¯ï¼š{secret}")
    print("ğŸ’ª åˆ¥æ°£é¤’ï¼Œå†è©¦ä¸€æ¬¡å§ï¼")
    print("=" * 50)


def play_game(min_num, max_num, max_attempts):
    """
    æ ¸å¿ƒéŠæˆ²é‚è¼¯
    
    åƒæ•¸ï¼š
        min_num: æœ€å°å€¼
        max_num: æœ€å¤§å€¼
        max_attempts: æœ€å¤§çŒœæ¸¬æ¬¡æ•¸
    
    å›å‚³ï¼š
        tuple: (won, attempts) - (æ˜¯å¦å‹åˆ©, çŒœæ¸¬æ¬¡æ•¸)
    """
    # ç”¢ç”Ÿéš¨æ©Ÿç­”æ¡ˆ
    secret_number = random.randint(min_num, max_num)
    attempts = 0
    guess_history = []
    
    # éŠæˆ²é–‹å§‹è¨Šæ¯
    print("\n" + "=" * 50)
    print("       ğŸ® éŠæˆ²é–‹å§‹ Game Start       ")
    print("=" * 50)
    print(f"ç¯„åœï¼š{min_num}-{max_num}ï¼Œä½ æœ‰ {max_attempts} æ¬¡æ©Ÿæœƒï¼")
    print("è«‹çŒœå‡ºé›»è…¦æƒ³çš„æ•¸å­—ï¼")
    
    # éŠæˆ²ä¸»è¿´åœˆ
    while attempts < max_attempts:
        attempts += 1
        
        # å–å¾—æœ‰æ•ˆçŒœæ¸¬
        guess = get_valid_guess(min_num, max_num, attempts, max_attempts)
        
        # è¨˜éŒ„æ­·å²
        guess_history.append(guess)
        
        # æª¢æŸ¥çµæœ
        result = check_guess(guess, secret_number)
        
        if result == 'correct':
            # å‹åˆ©
            show_guess_history(guess_history)
            show_victory_message(attempts)
            return (True, attempts)
        elif result == 'too_high':
            print("ğŸ“‰ å¤ªå¤§äº†ï¼")
        else:  # too_low
            print("ğŸ“ˆ å¤ªå°äº†ï¼")
        
        # é¡¯ç¤ºçŒœæ¸¬æ­·å²
        show_guess_history(guess_history)
    
    # éŠæˆ²çµæŸï¼ˆå¤±æ•—ï¼‰
    show_defeat_message(secret_number, max_attempts)
    return (False, attempts)


def ask_play_again():
    """
    è©¢å•æ˜¯å¦å†ç©ä¸€æ¬¡
    
    å›å‚³ï¼š
        bool: True=å†ç©, False=çµæŸ
    """
    print("\n" + "-" * 50)
    while True:
        choice = input("\nğŸ”„ å†ç©ä¸€æ¬¡ï¼Ÿ(y/n): ").strip().lower()
        
        if choice in ['y', 'yes', 'æ˜¯']:
            print("\nâœ… é–‹å§‹æ–°éŠæˆ²ï¼")
            return True
        elif choice in ['n', 'no', 'å¦']:
            print("\nğŸ‘‹ æ„Ÿè¬éŠç©ï¼")
            return False
        else:
            print("âŒ è«‹è¼¸å…¥ y æˆ– nï¼")


def show_final_stats(total_games, total_wins, attempts_list):
    """
    é¡¯ç¤ºæœ€çµ‚éŠæˆ²çµ±è¨ˆ
    
    åƒæ•¸ï¼š
        total_games: ç¸½éŠæˆ²æ¬¡æ•¸
        total_wins: å‹åˆ©æ¬¡æ•¸
        attempts_list: æ¯å±€çŒœæ¸¬æ¬¡æ•¸åˆ—è¡¨
    """
    print("\n" + "=" * 50)
    print("       ğŸ“Š éŠæˆ²çµ±è¨ˆ Game Statistics       ")
    print("=" * 50)
    print(f"ğŸ® ç¸½éŠæˆ²æ¬¡æ•¸ï¼š{total_games} å±€")
    print(f"ğŸ† å‹åˆ©æ¬¡æ•¸ï¼š{total_wins} å±€")
    print(f"ğŸ˜¢ å¤±æ•—æ¬¡æ•¸ï¼š{total_games - total_wins} å±€")
    
    if total_games > 0:
        win_rate = (total_wins / total_games) * 100
        print(f"ğŸ“ˆ å‹ç‡ï¼š{win_rate:.2f}%")
    
    if attempts_list:
        avg_attempts = sum(attempts_list) / len(attempts_list)
        print(f"ğŸ“Š å¹³å‡çŒœæ¸¬æ¬¡æ•¸ï¼š{avg_attempts:.2f} æ¬¡")
        print(f"â­ æœ€ä½³ç´€éŒ„ï¼š{min(attempts_list)} æ¬¡")
    
    print("=" * 50)
    print("\næ„Ÿè¬éŠç©çµ‚æ¥µå¯†ç¢¼éŠæˆ²ï¼")
    print("æ­¡è¿ä¸‹æ¬¡å†ä¾†æŒ‘æˆ°ï¼ğŸ‘‹")


def main():
    """
    ä¸»ç¨‹å¼
    """
    # éŠæˆ²æ¨™é¡Œ
    print("\n" + "=" * 60)
    print("     ğŸ® çµ‚æ¥µå¯†ç¢¼éŠæˆ² Ultimate Password Game ğŸ®     ")
    print("=" * 60)
    print("\næ­¡è¿ä¾†åˆ°çµ‚æ¥µå¯†ç¢¼éŠæˆ²ï¼")
    print("ä½¿ç”¨é‚è¼¯æ¨ç†ï¼Œåœ¨æœ‰é™æ¬¡æ•¸å…§æ‰¾å‡ºæ­£ç¢ºæ•¸å­—ï¼")
    
    # çµ±è¨ˆè®Šæ•¸
    total_games = 0
    total_wins = 0
    attempts_list = []  # è¨˜éŒ„æ¯å±€çŒœæ¸¬æ¬¡æ•¸
    
    # éŠæˆ²ä¸»è¿´åœˆ
    while True:
        # é¸æ“‡é›£åº¦
        min_num, max_num, max_attempts = choose_difficulty()
        
        # é€²è¡ŒéŠæˆ²
        won, attempts = play_game(min_num, max_num, max_attempts)
        
        # è¨˜éŒ„çµ±è¨ˆ
        total_games += 1
        if won:
            total_wins += 1
            attempts_list.append(attempts)
        
        # è©¢å•å†ç©ä¸€æ¬¡
        if not ask_play_again():
            break
    
    # é¡¯ç¤ºæœ€çµ‚çµ±è¨ˆ
    show_final_stats(total_games, total_wins, attempts_list)


# åŸ·è¡ŒéŠæˆ²
if __name__ == "__main__":
    main()
````
