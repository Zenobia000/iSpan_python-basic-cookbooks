## 💡 提示與技巧

### 除錯技巧
1. **使用 print() 追蹤變數**：
   ```python
   print(f"DEBUG: attempts={attempts}, guess={guess}")
   ```

2. **先測試小範圍**：
   ```python
   # 測試時使用小範圍
   secret_number = random.randint(1, 10)  # 1-10 容易測試
   ```

3. **固定答案測試**：
   ```python
   # 測試時先固定答案
   secret_number = 50  # 取消隨機
   ```

### 常見錯誤
1. **忘記更新計數器** → 無窮迴圈
2. **錯誤輸入計入次數** → 對玩家不公平
3. **範圍判斷錯誤** → 邊界值無法輸入
4. **沒有 break** → 答對後仍繼續
