## 🏆 進階挑戰

完成基本要求後，可以嘗試以下進階功能：

```python
# 進階挑戰 1：錯誤處理
# 處理除以零的情況

def safe_calculator():
    """包含錯誤處理的計算機"""
    print("=== 安全計算機 ===")
    
    # 取得輸入
    num1 = float(input("請輸入第一個數字: "))
    operator = input("請輸入運算符號 (+, -, *, /): ")
    num2 = float(input("請輸入第二個數字: "))
    
    # 執行運算（加入錯誤處理）
    if operator == "+":
        result = num1 + num2
    elif operator == "-":
        result = num1 - num2
    elif operator == "*":
        result = num1 * num2
    elif operator == "/":
        if num2 == 0:
            print("錯誤：不能除以零！")
            return
        result = num1 / num2
    else:
        print("錯誤：不支援的運算符號！")
        return
    
    # 顯示結果
    print(f"{num1} {operator} {num2} = {result}")

# 測試安全計算機
# safe_calculator()
```

```python
# 進階挑戰 2：更多運算符號
# 支援次方、餘數、整數除法

def advanced_calculator():
    """進階計算機"""
    print("=== 進階計算機 ===")
    print("支援運算：+, -, *, /, **, %, //")
    
    # 取得輸入
    num1 = float(input("請輸入第一個數字: "))
    operator = input("請輸入運算符號: ")
    num2 = float(input("請輸入第二個數字: "))
    
    # 執行運算
    if operator == "+":
        result = num1 + num2
    elif operator == "-":
        result = num1 - num2
    elif operator == "*":
        result = num1 * num2
    elif operator == "/":
        if num2 == 0:
            print("錯誤：不能除以零！")
            return
        result = num1 / num2
    elif operator == "**":
        result = num1 ** num2
    elif operator == "%":
        if num2 == 0:
            print("錯誤：不能對零求餘數！")
            return
        result = num1 % num2
    elif operator == "//":
        if num2 == 0:
            print("錯誤：不能除以零！")
            return
        result = num1 // num2
    else:
        print("錯誤：不支援的運算符號！")
        return
    
    # 顯示結果
    print(f"{num1} {operator} {num2} = {result}")

# 測試進階計算機
# advanced_calculator()
```
